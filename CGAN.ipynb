### CGAN(Concitional GAN)

#### Generative adversarial nets
GAN(Generative adversarial nets)は、生成モデル$G$と識別モデル$D$の2つの敵対的(adversarial)なモデルで構成されます。
$G$と$D$は同時に訓練され、$G$のパラメータは$\log (1-D(G(z)))$を最小化するように、$D$のパラメータは$\log D(x)$を最小化するように調整されます。
$$
\min_G\max_D L(D,G)=E_{x\sim p_{data}(x)}[\log D(x)]+E_{z\sim p_{z}(z)}[\log(1- D(G(z)))]
$$

#### Conditional Adversarial Nets
生成器と識別器にクラスラベル$\boldsymbol{y}$を与えることで、GANに条件つきのモデルに拡張できます。
生成器では、ノイズ$\boldsymbol{z}$にクラスを指定するワンホットベクトルを結合したノイズを入力とします
識別器では、ワンホットベクトルを画像サイズに拡大したクラス数分の条件画像をチャンネル方向に結合したものを使用します。(該当するクラスの要素が1、残りがすべて0の画像となります。)

損失関数は次のようになります。
$$
\min_G\max_D V(D,G)=E_{x\sim p_{data}(x|y)}[\log D(x)]+E_{z\sim p_{z}(z)}[\log(1- D(G(z|y)))]
$$
