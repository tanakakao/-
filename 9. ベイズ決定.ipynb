### ベイズの定理
ベイズ決定とは、ベイズの定理を用いた統計的決定である。
ベイズの定理は、原因の確率に関する定理である。いま、$k$個の可能なすべての原因を$H_1,\cdots,H_k$とする。
異なった原因$H_i,H_j$は同時には成立せず、互いに背反$H_i\cap H_j=\phi$としておく。
各$H_1,\cdots,H_k$の確率は、最初は$P(H_1),\cdots,P(H_k)$とする。
この確率は、あらかじめ予想、確信に対して数学的に簡単な数字を与えたもので、事前確率と呼ばれる。

いま、事象$A$を考えこれを結果とする。各原因からこの$A$が生じる条件付確率は、$P(A|H_1),\cdots,P(A|H_k)$であるとする。
ベイズの定理は、$A$を生じさせた各原因の可能性は、条件付確率として
$$
P(H_i|A)=\frac{P(H_i)P(A|H_i)}{\sum P(H_i)P(A|H_i)}
$$
によって計算される。この$P(H_i|A)$を$P(H_i)$に対して、事後確率と呼ぶ。事後確率には結果$A$が含まれているから、これをデータと読み替えれば、ベイズの定理に基づく統計学が成立する可能性があることがわかる。

「原因」として母数$\theta$をとり、「結果」として標本データ$z$をとり、さらに事前確率$P(H_i)$を$w(\theta_i)$、事後確率$P(H_i|A)$を$w'(\theta_i|z)$、さらに結果の確率分布$P(A|H_i)$を$f(z|\theta_i)$と表すと、$\theta$の事後確率
$$
w'(\theta_i|z)=\frac{w(\theta_i)f(z|\theta_i)}{\sum w(\theta_j)f(z|\theta_j)}
$$
を得る。$\theta$が連続的なら、和を積分にかえ
$$
w'(\theta|z)=\frac{w(\theta)f(z|\theta)}{\int_{\Theta} w(\theta)f(z|\theta)d\theta}
$$
となる。ここに$\Theta$は、可能な母数の値の集合で、母数空間と呼ばれる。
これら$w(\theta),w'(\theta|z)$を$\theta$の確率分布として見て、事前確率分布、事後確率分布という。
上の式は、$w,w'$および尤度関数$f(z|\theta)$の関係として、簡単に
$$
w'(\theta_i|z)\propto w(\theta_i)f(z|\theta)
$$
と書くことができる。

### 事前確率分布と事後確率分布
比較対照の試行で、新薬がプラセボに比して効くと判定されることを「成功」とすれば、$n$回の試行中$x$回の成功を得る確率は、二項分布
$$
f(x|\theta)={}_nC_x\theta^x(1-\theta)^{n-x},\hspace{3mm}x=0,1,\cdots,n
$$
で二項分布$Bi(n,\theta)$となる。
二項分布の形の尤度に対しては、その式の形に注目して$\theta$の事前確率分布としてベータ分布$Be(\alpha,\beta)$
$$
w(\theta)=\theta^{\alpha-1}(1-\theta)^{\beta-1} / B(\alpha, \beta),\hspace{3mm}0<\theta<1
$$
ただし、
$$
B(s,t)=\int_0^1u^{s-t}(1-u)^{t-1}du
$$
を採用すると、
$$
w'(\theta|x)\propto \theta^{\alpha+x-1}(1-\theta)^{\beta+(n-x)-1}
$$
であって、$\theta$の事後確率分布は$Be(\alpha+x,\beta+(n-x))$となる。
この場合、同じベータ分布の中で
$$
\alpha→\alpha+x,\hspace{5mm}\beta→\beta+(n-x)
$$
の変換をするだけでよい。
一般に、尤度$f(x|\theta)$に対して、$w(\cdot)$と$w'(\cdot|x)$が同一種類の分布(分布族)に属するならば、標本値$x$は単に同一分布族内の変換を引き起こすだけである。このとき、$w,w'$の分布族を$f$の自然な共役分布の族という。
二項分布$Bi(n,\theta)$の自然な共役分布は、ベータ分布$Be(\alpha,\beta)$である。

自然な共役分布でよく知られているのは。
1. 正規分布$N(\theta,\sigma^2)$の平均の母数$\theta$について、正規分布$N(\lambda,\tau^2)$
2. ポアソン分布$Po(\theta)$に対して、ガンマ分布$Ga(\alpha,\lambda)$
がそれぞれ自然共役分布となる場合である。
$z=(x_1,x_2,\cdots,x_n)$を正規母集団$N(\theta,\sigma^2)$からの、大きさ$n$の標本とするならば、尤度は
$$
\begin{align}
f(z|\theta)&=\prod_{i=1}^n\frac{1}{\sqrt{2\pi}\theta}\exp\bigg\{-\frac{(x_i-\theta)^2}{2\sigma^2} \bigg\}\\
&=\exp\bigg\{-\frac{(\bar{x}-\theta)^2}{2\sigma^2/n} \bigg\}\bigg(\frac{1}{\sqrt{2\pi}\sigma} \bigg)^n\exp\bigg\{-\frac{\sum(x_i-\bar{x})^2}{2\sigma^2} \bigg\}
\end{align}
$$
となり、$\theta$の関数は第一因数だけとなる。
第一因数中の統計量$\bar{x}$を知れば、$\theta$を知るためには十分である。
$$
w'(\theta|z)\propto w(\theta)\exp\bigg\{-\frac{(\bar{x}-\theta)^2}{2\sigma^2} \bigg\}
$$
となる。この式を見れば、$w(\theta)$も正規分布であって
$$
w(\theta)=\frac{1}{\sqrt{2\pi}\tau}\exp\bigg\{-\frac{(\theta-\lambda)^2}{2\tau^2} \bigg\}
$$
ととれば、計算上好都合であることがわかる。
$$
w'(\theta|z)\propto \exp\bigg\{-\frac{(\theta-\lambda')^2}{2\tau'^2} \bigg\}
$$
となる。ここで
$$
\begin{align}
1/\tau'^2&=(1/\tau^2)+(n/\sigma^2)\\
\lambda'&=\frac{(1/\tau^2)\lambda+(n/\sigma^2)\bar{x}}{(1/\tau^2)+(n/\sigma^2)}
\end{align}
$$
である。標本の大きさ$n$が大きくなるほど、$\lambda'$においては、標本平均$\bar{x}$が大きく効き、$\lambda'$は$\bar{x}$に引き付けられる。
ここで、分散の逆数$h=1/\tau^2,h_s=a/\sigma^2$を精度と呼ぶ。

### ベイズ推定
一般に、事後確率分布は、母数$\theta$の確率分布を示すことにより、$\theta$のありかの予想を示したものである。
$\theta$の値として一つの値($a$とする)を示すことは一つの決定であり、統計学的には推定の問題となる。
事後確率分布$w'(\theta|z)$から、$\theta$を推定する$a$の値を、$z$の関数(推定値)として取り出す方法を考える。
一般に$a$と$\theta$はなるべく近いほうがよい。そこでこの推定のペナルティ(損失関数)を$L(\theta,a)$とおき、これを基準に考える。

#### 期待損失の最小化の原理
$L(\theta,a)$の期待値(平均値)
$$
\sum_zf(z|\theta_i)\cdot L(\theta_i,a(z))
$$
または
$$
\int f(z|\theta)\cdot L(\theta,a(z))dz
$$
を最小にする$z$の関数$a(z)$として、最適な行動を求める。
$\theta$はもともと$w(\theta_i)$という確率分布をもっているのであるから、さらに期待値をとった
$$
\sum_iw(\theta_i)\{\sum_zf(z|\theta_i)\cdot L(\theta_i,a(z))\}
$$
が、推定の基準となる。
ベイズの定理を用いれば、この最小化は、$z$ごとに
$$
\sum_iw'(\theta_i|z)\cdot L(\theta_i,a(z))
$$
を最小にすることに帰する。
すなわち、事後確率分布による損失関数の期待値(期待損失)を最小化すればよいことがわかる。

#### ベイズ推定
事後確率分布による損失$L(\theta,a)$の期待値を最小にする$a(z)$を実際に求める。
1. 絶対損失$L(\theta,a)=|\theta-a|$  
一般に確率変数$X$に対する最小化
$$
\min_aE|X-a|
$$
の解は、$a(z)=$($X$のメディアン)であるから
$$
a(z)=(事後確率分布w'(\theta|z)のメディアン)
$$
となる。
2. 平方損失$L(\theta,a)=(\theta-a)^2$
一般に確率変数$X$に対する最小化
$$
\min_aE(X-a)^2
$$
の解は、$a=E(X)$であるから
$$
a(z)=(事後確率分布w'(\theta|z)の期待値)
$$
となる。
3. 単純損失
$\Delta>0$が十分小さい定数で
$$
L(\theta,a)=\left\{\begin{array}{ll}
0&|\theta-a|\leq \Delta\\
1&|\theta-a|> \Delta
\end{array}\right.
$$
この場合には、$a$を中心とする$±\Delta$の区間$[a-\Delta,a+\Delta]$に最大の事後確率が入ればよいから
$$
a(z)=(事後確率分布w'(\theta|z)のモード)
$$
となる。
4. 非対称絶対損失
1.の場合を非対称にして
$$
L(\theta,a)=\left\{\begin{array}{ll}
k_0|\theta-a|&a\geq \theta\\
k_1|\theta-a|&a< \theta
\end{array}\right.
$$
過大推定と過小推定を分ける場合に相当する。結果は$c=k_1/(k_0+k_1)$として、
$$
a(z)=(事後確率分布w'(\theta|z)の100cパーセント点)
$$

#### 二項分布のθのベイズ推定
事前確率分布、事後確率分布がベータ分布のときには、標本データ$z$に基づくベイズ推定$a(z)$(これを一般に$\hat{\theta}$と表す)は、平方損失の場合
$$
\hat{\theta}=\frac{\alpha+x}{\alpha+\beta+n}
$$
であり、単純損失の場合、$Be(\alpha,\beta)$のモードが
$$
M=\frac{\alpha-1}{\alpha+\beta-2}
$$
であるから
$$
\hat{\theta}=\frac{\alpha+x-1}{\alpha+\beta+n-2}
$$
となる。

#### 正規分布の平均θのベイズ推定
正規母集団$N(\theta,\sigma^2)$からの標本データ$z=(x_1,\cdots,x_n)$にもとづいて、それぞれ絶対、平方、単純の各損失を用いるベイズ推定$\hat{\theta}$は、正規分布のメディアン、平均、モードはすべて一致することから
$$
\hat{\theta}=\frac{(1/\tau^2)\lambda+(n/\sigma^2)\bar{x}}{(1/\tau^2)+(n/\sigma^2)}
$$
で与えらえる。ここに、$\lambda$は事前確率分布の平均、$\tau^2$は同じく分散、$\sigma^2$は母分散、$\bar{x}$は確率平均$\bar{x}=(x_1+\cdots+x_n)/n$である。

ここでは、$n$の値は固定してある。これを固定標本数の場合という。
$n=t$、$\lambda$を$\hat{\theta}_t$、$\lambda'$を$\hat{\theta}_{t+1}$、$x=x_{t+1}$とおく。
この記号を用いて書き直すと、
$$
\begin{align}
\hat{\theta}_{t+1}&=\frac{(1/\tau^2)\hat{\theta}_t+(n/\sigma^2)x_{t+1}}{(1/\tau^2)+(n/\sigma^2)}\\
&=\frac{\sigma^2\hat{\theta}_t+\tau^2x_{t+1}}{\sigma^2+\tau^2}
\end{align}
$$
となる。

### 統計的決定理論
#### 推定と検定の一般化
問題や場合、状況にふさわしい基準を採用することにより、その場合の最適な判断を決めることができる。
推定も検定もある基準になるコストを最小にする統計量を求めている、という一般的見方ができる。
統計的推測全体を一つの見地から統一的にとらえるこの考え方を、ワルドの統計的決定理論という。
また、ここでコストを表す関数が、一般の損失関数である。

母集団の未知の母数$\theta$に対する、ある判断の行動$a$の正しさの評価が問題となる。
推定では$a$は推定値(を与えるという行動)、検定では$a$は採択、棄却の行動である。
損失関数は$\theta$と$a$で決まるので、$L(\theta,a)$と書かれる。

統計的決定理論は、損失関数$L(\theta,a)$を基準にして、最適な行動$a$を、データ$z$を元にして決定するという一般的考え方のことである。

#### 統計的決定問題
統計的決定理論は、最適の行動$a$を、データの関数
$$
a=a(z)
$$
として定めることを目的とする。これを統計的決定理論という。この一般的な解法はベイズ推定を例に用いたように、最小事後期待損失の基準である。
