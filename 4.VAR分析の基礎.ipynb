### VAR分析の基本的なメリットとデメリット
VARモデルは自己回帰型の連立方程式モデルであり、相互依存関係にある複数の内生変数を、それらのラグ付き内生変数に関係づける基本構造を持っている。
内生変数が$K$個ある場合は、$K$本の連立方程式からなるモデルを構築する。
メリットとしては、予測が容易であり、予測精度が高いことが経験的に知られている。
デメリットとして、まず、ある内生変数のラグ付き内生変数が数個含まれている場合、これらの係数から意味のある情報を得ることが困難である。
また、推定すべき係数パラメータの個数が多くなり、標本サイズが小さいと、自由度が尽きてしまい、係数推定量の標準誤差が大きくなる。

### ものごとの直交分解
興味。分析の対象物となる量的な確率変数$y$が、2つの確率変数$f$と$u$に分解でき、次の構図になっているとする。
$$
y=f+u
$$
これらの3つの確率変数には次の関係式が成立する。
$$
\begin{align}
Var(y)&=Var(f+u)\\
&=Var(f)+Var(u)+2Cov(f,u)
\end{align}
$$
もし、$f$と$u$が無相関であれば、$Cov(f,u)=0$であり、次の関係式が成立する。
$$
\begin{align}
Var(y)&=Var(f+u)\\
&=Var(f)+Var(u)
\end{align}
$$
このタイプの分割は、直交分解という。
直交分解の関係式は、次のように表現することもできる。
$$
1=\frac{Var(f)}{Var(y)}+\frac{Var(u)}{Var(y)}
$$
右の2要素は$y$の動きを説明する寄与度を表す。

### 直交分解から見たインパルス応答分解
外生的なショックが経済的にどのような影響をもたらすかといった分析は、計量分析の主要なトピックである。
2分割化の延長線上に立てば、VARモデルは次の構図になっている。
$$
y=f_y+f_z+u
$$
ただし、$y$は内生変数のベクトル、$f_y$はラグ付き内生変数からなる部分、$f_z$は外生変数や定数項からなる部分、$u$は撹乱項のベクトルを表す。
標準的な過程に従い、$f_z$と$u$とは無相関とする。
伝統的な連立方程式モデルに基づく乗数分析では外生変数が外生的なショックの震源地となり、その波及効果は次のように表現できる。
$$
\Delta y=\Delta f_z(ショック発生時)\\
\Delta y=\Delta f_y(それ以降)
$$
ただし、$\Delta f_y$は$f_y$の変化分であり、ラグ付き内生変数を$y_{LAG}$と表記すれば、
$$
\Delta f_y=f(\Delta y_{LAG})
$$
といった変化分である。
VARモデルでは撹乱項が外生的なショックの震源地となり、その波及効果は次のように表現できる。
$$
\Delta y=\Delta u(ショック発生時)\\
\Delta y=\Delta f_y(それ以降)
$$

いずれも、ショック発生後は$\Delta f_y=f(\Delta y_{LAG})$と$\Delta y=\Delta f_y$を用い、$\Delta y$が次々に計算できる。

VARモデルは現在の内生変数が次期のラグ付き内生変数に変身する自己回帰構造になっているので、第1期の方程式は第2期の方程式に挿入でき、第2期の方程式は第3期の方程式に挿入できる。つまり、MA($\infty$)表現のベクトル版を得ることを行う。そのベクトル版は$f_y$が$u$の線形関数となり、ラグ付き内生変数を含まない表現になる。その変換後のモデルを便宜上、次のように表現する。
$$
y=g(u)+f_z
$$
ただし、$g(u)$は$f_y$と$u$を一体化した部分を表し、$u$の変動を線形変換するメカニズムになっている。
直交分解の結果、$u$に生じたショックの波及効果(限界効果)は次の形で表現できる。
$$
\Delta y=g(\Delta u)
$$
ただし、$g(\Delta u)$はショック$\Delta u$に起因する$g(u)$の変化分である。
自己回帰構造に組み込まれた撹乱項$u$の時系列は「イノベーション」と呼ばれる。

### 標準的なVARモデル
分析対象とする$K$個の量的変数の$t$期の値からなるベクトルを、
$$
\boldsymbol{y}_t=(y_{1t},y_{2t},\cdots,y_{Kt})'
$$
とする。これらの変数は同時点で相互に依存するとともに過去の互いの値にも依存し、さらにシステム外からの外生的なショックの影響を受けて変動しているものとする。それらの依存関係は近似的に線形で次のように表すことができるとする。
$$
\boldsymbol{A}\boldsymbol{y}_t=\boldsymbol{A}\boldsymbol{y}_{t-1}+\cdots+\boldsymbol{A}_p\boldsymbol{y}_{t-p}+\boldsymbol{A}_z\boldsymbol{z}_t+\boldsymbol{\varepsilon}_t\\
t=1,2,\cdots,T
$$
ここで、$\boldsymbol{z}_t$は外生変数や定数項を含むベクトルであり、$\boldsymbol{z}_t=(z_{1t},\cdots,z_{qt})'$とする。
$\boldsymbol{A}$および$\boldsymbol{A}_i(i=1,2,\cdots,p)$は$K×K$の係数行列、$\boldsymbol{A}_z$は$K×q$の係数行列、$\boldsymbol{\varepsilon}_t$は$K$個の撹乱項からなるベクトルである。
各方程式に含まれる係数パラメータの個数は$n=K×p+q$であり、$K$本の方程式を対象とする係数パラメータの総個数は$N=K×n$になる。
$\boldsymbol{z}_t$は雑多な要素を入れる役割があり、定数項、タイムトレンド項、季節ダミー変数等々を含めるためのベクトルである。
VARモデルにおいて、定数項やタイムトレンド項は確定項と呼ばれることから、$\boldsymbol{z}_t$は確定項ベクトルと呼ぶことにする。
上の定式化は理論的な制約を加えたVARモデルであり、連立方程式モデルの構造型に相当することから、構造VARモデル(SVARモデル)と呼ばれる。ラグ次数$p$の構造VARモデルといった意味でSVAR(p)と表記する。

一方、構造型に対する誘導型は次のようになる。
$$
\boldsymbol{y}_t=\boldsymbol{B}_1\boldsymbol{y}_{t-1}+\boldsymbol{B}_2\boldsymbol{y}_{t-2}+\cdots+\boldsymbol{B}_p\boldsymbol{y}_{t-p}+\boldsymbol{B}_z\boldsymbol{z}_t+\boldsymbol{u}_t\\
t=1,2,\cdots,T
$$
ただし、
$$
\boldsymbol{B}_i=\boldsymbol{A}^{-1}\boldsymbol{A}_i,i=1,2,\cdots,p\\
\boldsymbol{B}_z=\boldsymbol{A}^{-1}\boldsymbol{A}_z\\
\boldsymbol{u}_t=\boldsymbol{A}^{-1}\boldsymbol{\varepsilon}_1
$$

である。
この構造のVARモデルは、誘導VARモデルと呼ばれる。単にVARモデルと呼ばれることが多い。
ラグ次数$p$のVARモデルといった意味でVAR(p)と表記する。
誘導撹乱項ベクトル$\boldsymbol{u}_t$は次の性質を持つ定常過程とする。
$$
\begin{align}
E(\boldsymbol{u}_t)&=0,\ t=1,2,\cdots,T\\
Var(\boldsymbol{u}_t)&=E(\boldsymbol{u}_t\boldsymbol{u}_t')\\
&=\boldsymbol{\Sigma}_u\\
&=\left(\begin{array}{cccc}
\sigma_{11}&\sigma_{12}&\cdots&\sigma_{1K}\\
\sigma_{21}&\sigma_{22}&\cdots&\sigma_{2K}\\
\vdots&\vdots&&\vdots\\
\sigma_{K1}&\sigma_{K2}&\cdots&\sigma_{KK}
\end{array}
\right)\\
t=1,2,\cdots,T\\
Cov(\boldsymbol{u}_t,\boldsymbol{u}_s)&=E(\boldsymbol{u}_t\boldsymbol{u}_s')=0,\ t\neq s
\end{align}
$$
ただし、$\boldsymbol{\Sigma}_u$は$K×K$の対称行列であり、その$(j,j)$要素になる分散$\sigma_{jj}$は$\sigma_j^2$と表記することもある。
誘導型$\boldsymbol{\Sigma}_u$と構造型$\boldsymbol{\Sigma}_\varepsilon$との間には次の関係がある。
$$
\boldsymbol{\Sigma}_u=\boldsymbol{A}^{-1}\boldsymbol{\Sigma}_\varepsilon\bigl(\boldsymbol{A}^{-1}\bigr)'
$$
そして構造撹乱項ベクトル$\varepsilon_t$は次の性質を持つ定常過程とする。
$$
\begin{align}
E(\varepsilon_t)&=0,\ t=1,2,\cdots,T\\
Var(\varepsilon_t)&=E(\varepsilon_t\varepsilon_t')\\
&=\boldsymbol{\Sigma}_\varepsilon\\
&=\left(\begin{array}{cccc}
\sigma^\varepsilon_{11}&0&\cdots&0\\
0&\sigma^\varepsilon_{22}&\cdots&0\\
\vdots&\vdots&&\vdots\\
0&0&\cdots&\sigma^\varepsilon_{KK}
\end{array}
\right)\\
t=1,2,\cdots,T\\
Cov(\varepsilon_t,\varepsilon_s)&=E(\varepsilon_t\varepsilon_s')=0,\ t\neq s
\end{align}
$$

分散($Var(...)$)の式に示す性質は「均一分散」と呼ばれ、時間の変化に対して共分散行列が均一であることを意味する。
一方、$Cov(...)$の式は「系列無相関」と呼ばれ、異時点間の撹乱項が無相関であることを意味している。
VAR分析において、外生的な変化は$\boldsymbol{z}_t$で発生するのではなく、$\varepsilon_t$や$\boldsymbol{u}_t$で発生するのが基本になる。
$\varepsilon_t$や$\boldsymbol{u}_t$は外生的変化(ショック)の震源地といった意味で重要であり、ショックと呼ばれる場面がある。
さらに内生変数の変動を説明する革新的な過程といった意味でイノベーションと呼ばれる場面もある。

構造型の共分散行列$\boldsymbol{\Sigma}_\varepsilon$と係数行列$\boldsymbol{A}$に関しては次の2つの定式化がポピュラーである。

1. 係数行列$\boldsymbol{A}$の対角要素を1とする標準化を行い、共分散行列$\boldsymbol{\Sigma}$の対角要素(分散)を自由パラメータとする定式化がある。つまり、$Var(\varepsilon_t)=\boldsymbol{\Sigma}_\varepsilon$は$K×K$の対角行列とする定式化である
2. 共分散行列$\boldsymbol{\Sigma}_\varepsilon$の対角要素を1とする標準化を行い、係数行列$\boldsymbol{A}$の対角要素を自由パラメータとする定式化がある。このタイプの構造撹乱項は標準化されたものであることから、$\varepsilon_t^{SD}$と表記する。つまり、$Var(\varepsilon_t^{SD})=\boldsymbol{I}_K$とする定式化である。ただし、$\boldsymbol{I}_K$は$K×K$の単位行列である。

ラグ次数$p$はすべての方程式で同じ次数にするのが普通である。

```python
import numpy as np
import matplotlib.pyplot as plt

B = np.array([[0.1,-0.2],[0.5,0.1]])

fig, ax = plt.subplots(1, 2, figsize=(16,4))
y0 = 0
y1 = np.array([[y0],[y0]])

for _ in range(200):
    u = np.random.normal(0,1,size=(2,1))
    y1_next = B@y1[:,-1].reshape(-1,1) + u
    y1 = np.append(y1, y1_next,axis=1)
ax[0].plot(y1[0,:])
ax[1].plot(y1[1,:])
```

### ホワイトノイズ・ベクトル
定義は複数存在する。次の4つのベクトル過程を考える。

||過程(1)|過程(2)|過程(3)|過程(4)|
|-|-|-|-|-|
|異時点間における要素間の関係性|独立|独立|無相関|無相関|
|同時点間における要素間の関係性|独立|制約なし|無相関|制約なし|

ここでは過程(4)をホワイトノイズ・ベクトルと呼ぶ。
誘導撹乱項ベクトルはホワイトノイズ・ベクトルとなり、次のように表記する。
$$
\boldsymbol{u}_t\sim WN(\boldsymbol{0},\boldsymbol{\Sigma}_\varepsilon)\ または\ \boldsymbol{u}_t\sim VWN(\boldsymbol{0},\boldsymbol{\Sigma}_u)
$$

構造錯乱項ベクトル$\varepsilon_t$や$\varepsilon_t^{SD}$もホワイトノイズ・ベクトルとなり、次のように表記する。
$$
\varepsilon_t\sim WN(\boldsymbol{0},\boldsymbol{\Sigma}_\varepsilon)\\
\varepsilon_t^{SD} \sim WN(\boldsymbol{0},\boldsymbol{I}_K)
$$

一方、無相関に代えて独立を用いる場合は「独立ホワイトノイズ・ベクトル」と呼び、区別する。

### ベクトル過程の定常条件
ベクトル過程$\boldsymbol{y}_t=(y_{1t},y_{2t},\cdots,y_{Kt})'$に関し、その1次モーメントと2次モーメントとが有限で不偏であるとき、過程$\boldsymbol{y}_t$は弱定常や共分散定常という。
つまり、次の3つの条件を満たすベクトル過程$\boldsymbol{y}_t$は弱定常や共分散定常という。
$$
E(\boldsymbol{y}_t)=\boldsymbol{\mu}\\
E\bigl[(\boldsymbol{y}_t-\boldsymbol{\mu})(y_t-\mu)'\bigr]=\boldsymbol{\Sigma}_y\\
E\bigl[(\boldsymbol{y}_t-\boldsymbol{\mu})(y_{t-s}-\mu)'\bigr]=\boldsymbol{\Gamma}_y(s),\ s>0
$$

いずれも測度も時点$t$に依存しないことがポイントになっている。弱定常は定常と呼ぶのが普通である。
ここでは次の標準的なVAR(p)過程を対象とし、その定常条件を検討する。
$$
\boldsymbol{y}_t=\boldsymbol{B}_1\boldsymbol{y}_{t-1}+\boldsymbol{B}_2\boldsymbol{y}_{t-2}+\cdots+\boldsymbol{B}_p\boldsymbol{y}_{t-p}+\boldsymbol{u}_t
$$
ただし、$\boldsymbol{u}_t$はホワイトノイズ・ベクトルであり、それは定常なベクトル過程である。
このVAR(p)におけるラグ変数を方程式左辺へ移動させて次の方程式を得る。
$$
\boldsymbol{y}_t-\boldsymbol{B}_1\boldsymbol{y}_{t-1}-\boldsymbol{B}_2\boldsymbol{y}_{t-2}-\cdots-\boldsymbol{B}_p\boldsymbol{y}_{t-p}=\boldsymbol{u}_t
$$
これは非同次方程式と呼ばれるタイプになる。これに対する同次方程式は次のようになる。
$$
\boldsymbol{y}_t-\boldsymbol{B}_1\boldsymbol{y}_{t-1}-\boldsymbol{B}_2\boldsymbol{y}_{t-2}-\cdots-\boldsymbol{B}_p\boldsymbol{y}_{t-p}=0
$$
フィルターを使うと、同次方程式は次のように表現することができる。
$$
\boldsymbol{B}(L)\boldsymbol{y}_t=0
$$
ただし、
$$
\boldsymbol{B}(L)=\boldsymbol{I}_K-\boldsymbol{B}_1L^1-\boldsymbol{B}_2L^2-\cdots-\boldsymbol{B}_pL^p
$$

である。$\boldsymbol{B}(L)$は多変量フィルターと呼ぶ。
$\boldsymbol{B}(L)y_t$に示すVAR過程が定常か否かは、係数行列$(\boldsymbol{B}_1,\boldsymbol{B}_2,\cdots,\boldsymbol{B}_p)$からなる線形変換メカニズムの特性に関わっている。
VAR(1)形式に変換し、その線形変換メカニズムの行列の特性を調べることを行う。
VAR過程が定常か否かは、次の固有方程式で判別できる。
$$
\det (\boldsymbol{I}_K\lambda^p-\boldsymbol{B}_1\lambda^{p-1}-\boldsymbol{B}_2\lambda^{p-2}-\cdots-\boldsymbol{B}_p)=0
$$
ただし、$\lambda$は固有値である。この固有方程式を解く個々の固有値$\lambda_i$のサイズが次の条件を満たしていれば、当該VAR過程は定常である。
$$
|\lambda_i|<1
$$
それ以外、つまり
$$
|\lambda_i|\geq1
$$
を満たす固有値が1個でもあれば当該VAR過程は非定常である。
$\lambda$としては複素数の可能性があるので、$|\lambda|$は複素数のサイズを測るモジュラスである。
次のラグ型特性方程式を使っても判別できる。
$$
\det(\boldsymbol{I}_K-\boldsymbol{B}_1z-\boldsymbol{B}_2z^2-\cdots-\boldsymbol{B}_pz^p)=0
$$
ラグ型特性方程式を解く個々の$z_i$が$|z_i|>1$であれば、当該VAR過程は定常である。それ以外、当該VAR過程は非定常である。

確定項ベクトルをVAR過程に含める場合でも、前述したI(d)過程の線形結合に関する性質が役立つ。

1. 定常=定常+定常
2. 非定常=非定常+定常

これはVAR過程全体が定常になるためにはすべての要素が定常である必要性を表している。

```python
eig_values, eig_vactors = np.linalg.eig(B)
np.abs(eig_values)
```

### 和分過程の線形結合
単純化のため、$\boldsymbol{y}_t=(y_{1t},y_{2t})'$と$\boldsymbol{\beta}=(\beta_{1},\beta_{2})'$を用い、
$$
\boldsymbol{\beta}'\boldsymbol{y}=\beta_1y_{1t}+\beta_2y_{2t}
$$
といった線形結合を考える。ただし、$\beta_1$と$\beta_2$は任意な定数である。
定常過程の線形結合は定常である。つまり、$y_{1t}\sim I(0)$と$y_{2t}\sim I(0)$である場合は次の結果が得られる。
$$
\beta_1y_{1t}+\beta_2y_{2t}\sim I(0)
$$
この場合は$\boldsymbol{y}_t=(y_{1t},y_{2t})'$を使ってVARモデルを構築する。これはレベルVARと呼ばれる分類になる。
このVARモデルにおける第1方程式に焦点を置き、和分の次数から見た構図は次のようになる。
$$
y_{1t}=\beta_1y_{1t-1}+\beta_2y_{2t-1}+u_{1t}\\
I(0)=I(0)+I(0)+I(0)
$$

ただし、標準的な過程に従い、$u_t$は定常過程I(0)としている。
第2方程式も同じ構図になっている。したがってVARモデル全体としても定常である。
次は異なった次数の組み合わせを考える。
具体的には$y_{1t}\sim I(0)$と$y_{2t}\sim I(1)$となっているケースを考える。この場合の線形結合は次の結果になる。
$$
\beta_1y_{1t}+\beta_2y_{2t}\sim I(1)
$$
つまり、次数が異なる線形結合は次数の高い方の過程になる。
したがって、ベクトル過程$\boldsymbol{y}_t=(y_{1t},y_{2t})'$はI(1)に分類される。
一方、次の線形結合にすれば定常過程になる。
$$
\beta_1y_{1t}+\beta_2\Delta y_{2t}\sim I(0)
$$
この場合は$\boldsymbol{y}_t=(y_{1t},\Delta y_{2t})'$を用いてVARモデルを構築する。そうすればVARモデルとしても定常である。
一方、$\boldsymbol{y}_t=(\Delta y_{1t},\Delta y_{2t})'$を用いるべきという意見もある。
$$
\beta_1\Delta y_{1t}+\beta_2\Delta y_{2t}\sim I(0)
$$
これは内生変数を階差型でそろえる考え方である。
次は、どちらも定常過程I(0)でなく、同じ次数の組み合わせを考える。
具体的には$y_{1t}\sim I(1)$と$y_{2t}\sim I(1)$となっている場合を考える。この場合は2つの可能性がある。
$$
\beta_1y_{1t}+\beta_2y_{2t}\sim I(1)\\
\beta_1y_{1t}+\beta_2y_{2t}\sim I(0)
$$
前者の場合は階差をとると、定常過程になる。
$$
\Delta \beta_1y_{1t}+\Delta \beta_2y_{2t}\sim I(0)
$$
したがって、$\boldsymbol{y}_t=(\Delta y_{1t},\Delta y_{2t})'$を使ってVARを構築すればよい。
これは階差VARと呼ばれる分類になる。
後者の場合は共和分の関係が存在するケースになる。

```python
B = np.array([[1.001,-0.02],[0.05,0.1]])

fig, ax = plt.subplots(2, 2, figsize=(16,8))
y0 = 0
y1 = np.array([[y0],[y0]])

for _ in range(200):
    u = np.random.normal(0,1,size=(2,1))
    y1_next = B@y1[:,-1].reshape(-1,1) + u
    y1 = np.append(y1, y1_next,axis=1)

ax[0][0].plot(y1[0,:])
ax[0][1].plot(y1[1,:])
ax[1][0].plot(np.diff(y1[0,:]))
ax[1][1].plot(np.diff(y1[1,:]))
```

### 共和分関係と見せかけの回帰
2つの過程に関し、$y_{1t}\sim I(1)$と$y_{2t}\sim I(1)$でもその線形結合が定常過程I(0)になる場合がある。
つまり、
$$
\beta_1y_{1t}+\beta_2y_{2t}\sim I(0)
$$
となるケースがある。
このような関係を共和分の関係にあるという。
共和分の関係にある変数を回帰に使うと、次のことが起こる。
例えば次の線形回帰モデルを考える。
$$
y_{1t}=\beta_1+\beta_2y_{2t}+u_t
$$
ただし、$u_t$は定常な撹乱項とする。和分の次数から見た構図は次のようになっている。
$$
I(1)=I(1)+I(0)
$$
この場合、2変数$y_{1t},y_{2t}$が無関係であっても、回帰では次の症状が現れる。

1. 決定係数が大きい
2. t値が大きい
3. 残差の系列相関が大きい。したがってダービン=ワトソン検定統計量の値がゼロ近辺になり、系列相関なしを意味する2から大きくくずれる。

本来は関係のない変数であっても、回帰の結果から関係ありと判断する傾向がある。
このケースは見せかけの回帰と呼ばれる。

### 共和分関係の解釈
経済学では、「共和分の関係=均衡の関係」と解釈するのが普通である。

同じI(1)を持った時系列過程を線形結合すると、波長の合った関係が成立することがあり、線形結合が定常的になることがある。
したがって、「波長の合った関係=共和分の関係」と解釈できる。
波長の合った関係は、変数間の均衡関係を記述する表現としても適切である。さらに共和分が確率的な概念であることを考慮すると、「確率的に波長の合った関係=共和分の関係」といった解釈になる。

### 共和分の定義
$n$個の要素を含む次のベクトル過程を考える。
$$
\boldsymbol{y}_t=(y_{1t},y_{2t},\cdots,y_{nt})'
$$

さらに次に2つの条件を考える。

1. $\boldsymbol{y}_t$のすべての要素は、次数$d$の和分過程I(d)にある。
2. ベクトル$\boldsymbol{\beta}=(\beta_1,\beta_2,\cdots,\beta_n)'$に対し、$\boldsymbol{\beta}'\boldsymbol{y}_t=\beta_{1y1t}+\beta_2y_{2t}+\cdots+\beta_ny_{nt}$がI(d-b)となる関係が成立する。ただし、$b\geq1$の整数である。

これら2つの条件が満たされるとき、$\boldsymbol{y}_t$の要素は次数$(d,b)$で共和分しているという。
その関係は$\boldsymbol{y}_t\sim CI(d,b)$と表記し、$\boldsymbol{\beta}$は共和分ベクトルという。

独立した共和分ベクトルの個数は$\boldsymbol{y}_t$の共和分ランクという。

### 定常VARを目指して
「定常VAR」対「非定常VAR」といった対比は、一変量に関する「定常過程」対「単位根過程」の対比と同様である。

1. 定常VAR：I(0)に分類されるベクトル過程である。単位根を持っていないといった形で特徴づけが行われている。その意味は、固有値のスケールが1より小さいといった意味である。トレンド定常VARは定義的には非定常であるが、タイムトレンドの周りで定常であることから定常VARのグループに入れる。このグループはレベル定常VARとトレンド定常VARとに分けることができる。
2. 非定常VAR：I(1)やI(2)に分類されるベクトル過程である。単位根を持っているといった形で特徴づけが行われている。その意味は、固有値のスケールが1以上といった意味である。

推定の場面では次の構図が役立つ。

1. 定常VAR：単位根がないことに起因し、レベル型のままで伝統的な漸近理論が通用する。標準的な推定法が使える。
2. 非定常VAR：単位根があることに起因し、レベル型のままでは伝統的な漸近理論が通用しない。標準的な推定法が使えない。

推定の場面で目指す構図をイメージ的に示すと次のようになる。

内生変数=規則性で説明できるシステマティックな部分+それ以外のランダムな部分


推定に求められる環境については「モデル全体が定常になっているシステムにする」の一言に尽きる。

### ベクトル誤差修正モデル
$y_{1t}\sim I(1)$と$y_{2t}\sim I(1)$となっている場合は2つの可能性があった。
$$
\beta_1y_{1t}+\beta_2y_{2t}\sim I(1)\\
\beta_1y_{1t}+\beta_2y_{2t}\sim I(0)
$$

後者の場合が共和分関係のケースである。

共和分関係の場合、2変数$y_{1t},y_{2t}$からなるVAR(1)では次のような構図になる。
$$
y_{1t}=\beta_1y_{1t-1}+\beta_2y_{2t-1}+u_{1t}
$$
方程式左右の和分次数の構図は次のようになっている。
$$
I(1)=I(0)+I(0)
$$
一方、求めるのは次の構図である。
$$
I(0)=I(0)+I(0)
$$

さらに線形結合$(\beta_1y_{1t-1}+\beta_2y_{2t-1})$がI(0)過程であることは、$y_{1t}$と$y_{2t}$とが均衡関係にあると解釈できる。

単純化のために次のVAR(2)を考える。

$$
\boldsymbol{y}_t=\boldsymbol{B}_1\boldsymbol{y}_{t-1}+\boldsymbol{B}_2\boldsymbol{y}_{t-2}+\boldsymbol{u}_t
$$
式を整理すると、次の定式化が得らえる。
$$
\begin{align}
\boldsymbol{y}_t-\boldsymbol{y}_{t-1}=\boldsymbol{B}_1\boldsymbol{y}_{t-1}-\boldsymbol{y}_{t-1}+\boldsymbol{B}_2\boldsymbol{y}_{t-2}+\boldsymbol{u}_t\\
\Delta\boldsymbol{y}_t&=\boldsymbol{B}_1\boldsymbol{y}_{t-1}-\boldsymbol{y}_{t-1}+\boldsymbol{B}_2\boldsymbol{y}_{t-1}-\boldsymbol{B}_2\boldsymbol{y}_{t-1}+\boldsymbol{B}_2\boldsymbol{y}_{t-2}+\boldsymbol{u}_t\\
&=(\boldsymbol{B}_1+\boldsymbol{B}_2-\boldsymbol{I}_K)\boldsymbol{y}_{t-1}+(-\boldsymbol{B}_2)(\boldsymbol{y}_{t-1}-\boldsymbol{y}_{t-2})+\boldsymbol{u}_t\\
&=\Pi\boldsymbol{y}_{t-1}+\boldsymbol{\Gamma}_1\Delta\boldsymbol{y}_{t-1}+\boldsymbol{u}_t
\end{align}
$$
ただし、
$$
\boldsymbol{\Pi}=-\boldsymbol{I}_K+\boldsymbol{B}_1+\boldsymbol{B}_2,\hspace{3mm}\boldsymbol{\Gamma}_1=-\boldsymbol{B}_2
$$
である。
$\Delta\boldsymbol{y}_t$や$\Delta\boldsymbol{y}_{t-1}$はI(0)過程であり、レベル型線形結合$\boldsymbol{\Pi}\boldsymbol{y}_{t-1}$もI(0)過程である。したがって、目指す構図が成り立っている。
さらに係数行列は
$$
\boldsymbol{\Pi}=\boldsymbol{\alpha}\boldsymbol{\beta}'
$$
といった具合に分解できる。したがって次の定式化が成り立つ。
$$
\Delta\boldsymbol{y}_t=\boldsymbol{\alpha}\boldsymbol{\beta}\boldsymbol{y}_{t-1}+\boldsymbol{\Gamma}_1\Delta\boldsymbol{y}_{t-1}+\boldsymbol{u}_t
$$
こもモデルはベクトル誤差修正モデルと呼ばれ、VECモデルやVECM(1)と表記する。

(1)$\boldsymbol{\beta}'\boldsymbol{y}_{t-1}$は誤差修正項
(2)$\boldsymbol{\beta}$は共和分行列
(3)$\boldsymbol{\alpha}$は送出行列ないし調整行列

次に、ラグ次数$p$のVARを考える。
$$
\boldsymbol{y}_t=\boldsymbol{B}_1\boldsymbol{y}_{t-1}+\boldsymbol{B}_2\boldsymbol{y}_{t-2}+\cdots+\boldsymbol{B}_p\boldsymbol{y}_{t-p}+\boldsymbol{B}_z\boldsymbol{z}_{t}+\boldsymbol{u}_t
$$
同様なモデル変換を行って整理すると、次の定式化が得られる。
$$
\begin{align}
\Delta\boldsymbol{y}_t&=\boldsymbol{\Pi} \boldsymbol{y}_{t-1}+\boldsymbol{\Gamma}_1\Delta\boldsymbol{y}_{t-1}+\cdots+\boldsymbol{\Gamma}_{p-1}\Delta\boldsymbol{y}_{t-p+1}+\boldsymbol{B}_z\boldsymbol{z}_t+\boldsymbol{u}_t\\
&=\boldsymbol{\Pi} \boldsymbol{y}_{t-1}+\sum_{i=1}^{p-1}\boldsymbol{\Gamma}_i\Delta\boldsymbol{y}_{t-i}+\boldsymbol{B}_z\boldsymbol{z}_t+\boldsymbol{u}_t\\
&=\boldsymbol{\alpha}\boldsymbol{\beta}'\boldsymbol{y}_{t-1}+\sum_{i=1}^{p-1}\boldsymbol{\Gamma}_i\Delta\boldsymbol{y}_{t-i}+\boldsymbol{B}_z\boldsymbol{z}_t+\boldsymbol{u}_t
\end{align}
$$

ただし、
$$
\begin{align}
\boldsymbol{\Pi}&=-\boldsymbol{I}_K+\boldsymbol{B}_1+\cdots+\boldsymbol{B}_p\\
\boldsymbol{\Gamma}_i&=-(\boldsymbol{B}_{i+1}+\boldsymbol{B}_{i+2}+\cdots+\boldsymbol{B}_{p}),\ i=1,2,\cdots,p-1\\
\boldsymbol{\Pi} &= \boldsymbol{\alpha}\boldsymbol{\beta}'
\end{align}
$$

である。
これはラグ次数$p-1$のVECモデルであるからVEC(p-1)やVECM(p-1)と表記する。

共和分過程が誤差修正モデルとして表現できることは、グランジャーの表現定理と呼ばれる。

```python
B1 = np.array([[1.001,-0.001],[0.004, 1.001]])
B2 = np.array([[0.001,-0.002],[0.003, 0.001]])

Pi = -np.eye(len(B1))+B1+B2
Gamma1 = -B2

fig, ax = plt.subplots(4, 2, figsize=(24,10))
y0 = 0
y1 = np.array([[y0,y0],[y0,y0]])
y2 = np.array([[y0,y0],[y0,y0]])
delta_y = np.array([[y0],[y0]])

for _ in range(200):
    u = np.random.normal(0,1,size=(2,1))
    delta_y = np.append(delta_y, Pi@y2[:,-1].reshape(-1,1)+Gamma1@delta_y[:,-1].reshape(-1,1)+u,axis=1)
    y2 = np.append(y2, y2[:,-1].reshape(-1,1)+delta_y[:,-1].reshape(-1,1), axis=1)
    
    y1_next = B1@y1[:,-1].reshape(-1,1) + B2@y1[:,-2].reshape(-1,1) + u
    y1 = np.append(y1, y1_next,axis=1)
    
ax[0][0].plot(y1[0,:])
ax[0][1].plot(y1[1,:])
ax[1][0].plot(np.diff(y1[0,:]))
ax[1][1].plot(np.diff(y1[1,:]))

ax[2][0].plot(y2[0,:])
ax[2][1].plot(y2[1,:])
ax[3][0].plot(delta_y[0,:])
ax[3][1].plot(delta_y[1,:])
```

### 共和分過程における確定項
VECモデルにおける確定項の定式化については、数個のパターンが開発されている。
基本的には次の2つの考え方に集約することができる。
1. 共和分関係$\boldsymbol{\Pi} \boldsymbol{y}_{t-1}$の中に定数項やトレンド項を入れる
2. 共和分関係$\boldsymbol{\Pi} \boldsymbol{y}_{t-1}$の外側に定数項やトレンド項を入れる

タイムトレンドについては、
$$
\boldsymbol{c}_1+\boldsymbol{c}_2t
$$
を用いて考察する。
まず、タイムトレンドがある場合は、次の定式化が考えられる。
$$
\Delta \boldsymbol{y}_t = \boldsymbol{c}_1+\boldsymbol{\alpha}(\boldsymbol{\beta}'\boldsymbol{y}_{t-1}+\boldsymbol{c}_2t)+\cdots\\
\Delta \boldsymbol{y}_t = \boldsymbol{c}_1+\boldsymbol{c}_2t+\boldsymbol{\alpha}\boldsymbol{\beta}'\boldsymbol{y}_{t-1}+\cdots
$$
タイムトレンドがないが、定数項のある場合は、次の定式化が考えられる。
$$
\Delta \boldsymbol{y}_t = \boldsymbol{a}(\boldsymbol{\beta}'\boldsymbol{y}_{t-1}+\boldsymbol{c}_1)+\cdots\\
\Delta \boldsymbol{y}_t = \boldsymbol{c}_1+\boldsymbol{\alpha}\boldsymbol{\beta}'\boldsymbol{y}_{t-1}+\cdots
$$
タイムトレンドもなく、定数項もない場合は、次の定式化が考えられる。
$$
\Delta \boldsymbol{y}_t = \boldsymbol{\alpha}\boldsymbol{\beta}'\boldsymbol{y}_{t-1}+\cdots
$$
この定式化はすべての内生変数が同じゼロ平均を持つことを仮定する。

```python
B1 = np.array([[0.001,-0.001],[0.004, 0.001]])
B2 = np.array([[0.001,-0.002],[0.003, 0.001]])

c1 = np.array([[1],[3]])
c2 = np.array([[-0.05],[0.02]])

Pi = -np.eye(len(B1))+B1+B2
Gamma1 = -B2

fig, ax = plt.subplots(2, 2, figsize=(24,8))
y0 = 0
y1 = np.array([[y0,y0],[y0,y0]])

for t in range(200):
    u = np.random.normal(0,1,size=(2,1))
    delta_y = np.append(delta_y, Pi@y1[:,-1].reshape(-1,1)+Gamma1@delta_y[:,-1].reshape(-1,1)+c1+c2*t+u,axis=1)
    y1 = np.append(y1, y1[:,-1].reshape(-1,1)+delta_y[:,-1].reshape(-1,1), axis=1)
    
ax[0][0].plot(y1[0,:])
ax[0][1].plot(y1[1,:])
ax[1][0].plot(np.diff(y1[0,:]))
ax[1][1].plot(np.diff(y1[1,:]))
```

### VAR系モデルの分類
VAR系のモデルは「構造型」対「誘導型」といった形で分類できるが、「レベル型」対「階差型」といった視点から次の3タイプに分類することができる。

1. レベルVARモデル(level VAR)
2. 階差VARモデル(difference VAR)
3. ベクトル誤差修正モデル(VECまたはVECM)

レベル型は原系型と呼ばれることもある。
