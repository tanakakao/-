### ラグ演算子と階差演算子
1. リード演算子:$E_{y_t}\equiv y_{t+1}$
2. ラグ演算子:$L_{y_t}\equiv y_{t-1}$
3. 前方差分演算子:$\Delta y_t\equiv y_{t+1}-y_{t}$
4. 後方差分演算子:$\nabla y_t\equiv y_{t}-y_{t-1}$

時系列分析の分野では後方差分演算子がポピュラーであり、$\Delta y_t\equiv y_{t+1}-y_{t}$と定義し、階差演算子と呼ぶことが多い。
時系列分析では次の２つのタイプが一般的である。

1. ラグ演算子:$L_{y_t}\equiv y_{t-1}$
2. 階差演算子:$\Delta y_t\equiv y_{t}-y_{t-1}$

ラグ変換を2回行うと、次の結果が得られる。
$$
L(Ly_t)=Ly_{t-1}=y_{t-2}
$$
これは次のように表記する。
$$
L^2y_t=y_{t-2}
$$
任意の正の数$s$に関し、
$$
L^sy_t=y_{t-s}
$$
が成立し、$L^sy_t$は$s$次のラグ値を表す。
また、$s$は負の整数であってもよく、
$$
L^{-1}y_t=y_{t+1}
$$
が成立する。$L^{-1}$はリード演算子という。
ラグ演算子には次の性質がある。

$$
L(\beta y_t)=\beta×(Ly_t)\\
L(x_t+y_t)=Lx_t+Ly_t\\
(\beta_1 L+\beta_2 L^2)y_t=\beta_1 y_{t-1}+\beta_2 y_{t-2}
$$

これらの性質は$L$が線形変換であることに基づく。

階差変換を2回行うと、次の結果が得られる。

$$
\begin{align}
\Delta(\Delta y_t)&=\Delta(y_t-y_{t-1})\\
&=(y_t-y_{t-1})-(y_{t-1}-y_{t-2})\\
&=y_t-2y_{t-1}+y_{t-2}
\end{align}
$$

これは次のように表記する。

$$
\begin{align}
\Delta^2y_t&=\Delta(\Delta y_t)\\
&=y_t-2y_{t-1}+y_{t-2}
\end{align}
$$

そして$\Delta y_t$は1次の階差、$\Delta^2 y_t$は2次の階差と呼ぶ。
階差変換とラグ変換との間には次の関係がある。

$$
\Delta^d=(1-L)^d,\hspace{3mm}d=1,2,3,\cdots
$$

データの周期に対応し、次の時間差を選ぶこともある。

1. 四半期のデータ:$y_t-y_{t-4}$
2. 月周期のデータ:$y_t-y_{t-12}$
3. 日周期のデータ:$y_t-y_{t-7}$(同じ曜日のデータ間の階差)

季節調整済データに対して階差をとる場合は、次のことが起こるので注意する。
季節調整済データが次の移動平均で成り立っているとする。
$$
y_t^{ma}=\frac{y_t+y_{t-1}+y_{t-2}+y_{t-3}}{4}
$$
1次の階差をとると次のようになる。
$$
\begin{align}
\Delta y_t^{ma}&=\frac{(y_t-y_{t-1})+(y_{t-1}-y_{t-2})+(y_{t-2}-y_{t-3})+(y_{t-3}-y_{t-4})}{4}\\
&=\frac{y_t-y_{t-4}}{4}
\end{align}
$$
途中の値がすべて抜け落ち、多くの情報が失われた結果になる。

### データ生成過程
確率変数$y$の時系列データ$\{y_t\}_{t=1}^T$は確率変数列$\{y_t\}_{t=-\infty}^\infty$の1つの標本と位置付ける。
そして、確率変数列$\{y_t\}_{t=-\infty}^\infty$の方は確率過程やデータ生成過程(DGP)と呼ぶ。
確率過程は単に過程と呼ばれることが多く、$\{y_t\}_{t=-\infty}^\infty$は単に$y_t$と表記することが多い。
したがって、「過程$y_t$」と表記している場合は、標本の話ではなく、母集団の話をしているといった認識が必要である。

1. 過程$y_t$:理論(確率変数)の話をしている
2. $\{y_t\}_{t=1}^T$:データ(現実)の話をしている

### ホワイトノイズ
回帰分析における撹乱項や誤差項と呼ばれる確率変数は$u_t$や$\varepsilon_t$と表記することが多い。
時系列分析ではホワイトノイズとして特定化され、ノイズ項と呼ばれることが多い。
次の性質を持つ時系列過程$u_t$をホワイトノイズという。

$$
E(u_t)=0\\
Var(u_t)=E(u_tu_t)=\sigma^2,\hspace{3mm}\sigma^2<\infty\\
Cov(u_t,u_s)=E(u_tu_s)=0,\hspace{3mm}t\neq s
$$

ホワイトノイズは異時点間の自己相関がゼロである性質、言い換えると系列無相関である性質が特徴的である。
ホワイトノイズは次のように表記する。
$$
u_t\sim WN(0,\sigma^2)またはu_t\sim WN(\sigma^2)
$$
無相関に代えて独立を仮定するものは独立ホワイトノイズという。
したがって、独立ホワイトノイズは次のように表現できる。
$$
u_t\sim iid(0,\sigma^2)
$$
ただし、iidは「independently and identically distributed」の略称であり、「互いに独立で、かつ同一の分布に従う」を意味する。
「無相関」対「独立」の差異は「弱ホワイトノイズ」対「強ホワイトノイズ」といった形で区別することも行われている。
正規分布に従うホワイトノイズは正規ホワイトノイズという。次のように表記できる。
$$
u_t\sim iidN(0,\sigma^2)
$$
正規分布の下では「無相関=独立」が成立する。
### 代表的なデータ生成過程
代表的なデータ生成過程として、次の3つを挙げることができる。
1. 自己回帰過程
$$
y_t=c+\phi_1y_{t-1}+\phi_2y_{t-2}+\cdots+\phi_py_{t-p}+u_t,\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$
ただし、$c$は定数である。これは次数$p$の自己回帰過程と呼び、AR(p)と表記する。
ラグオペレータを用い、AR(p)は次のように表現することができる。
$$
(1-\phi_1L-\phi_2L^2-\cdots-\phi_pL^p)y_t=c+u_t\\
\phi(L)y_t=c+u_t
$$
ただし、
$$
\phi(L)=(1-\phi_1L-\phi_2L^2-\cdots-\phi_pL^p)
$$
2. 移動平均過程
$$
y_t=\mu+u_t+\theta_1u_{t-1}+\theta_2u_{t-2}+\cdots+\theta_qu_{t-q},\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$
これは次数$q$の移動平均過程と呼び、MA(q)と表記する。
ラグオペレータを用いて、MA(q)は次のように表現できる。
$$
\begin{align}
y_t&=\mu+(1+\theta_1L+\theta_2L^2+\cdots+\theta_qL^q)u_t\\
&=\mu+\theta(L)u_t
\end{align}
$$
ただし、
$$
\theta(L)=(1+\theta_1L+\theta_2L^2+\cdots+\theta_qL^q)
$$
3. 自己回帰移動平均過程
$$
y_t=c+\phi_1y_{t-1}+\phi_2y_{t-2}+\cdots+\phi_py_{t-p}+\theta_1u_{t-1}+\theta_2u_{t-2}+\cdots+\theta_qu_{t-q},\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$
これは次数$(p,q)$の自己回帰移動平均過程と呼び、ARMA(p,q)と表記する。
ラグオペレータを用い、ARMA(p,q)は次のように表現できる。
$$
(1-\phi_1L-\phi_2L^2-\cdots-\phi_pL^p)y_t=c+(1+\theta_1L+\theta_2L^2+\cdots+\theta_qL^q)u_t\\
\phi(L)y_t=c+\theta(L)u_t\\
\phi(L)(y_t-\mu)=\theta(L)u_t
$$
ただし、
$$
\phi(L)=(1-\phi_1L-\phi_2L^2-\cdots-\phi_pL^p)\\
\theta(L)=(1+\theta_1L+\theta_2L^2+\cdots+\theta_qL^q)\\
$$

$$
\begin{align}
\mu&=\frac{c}{\phi(L)}\\
&=\frac{c}{\phi(1)}\\
&=\frac{c}{1-\phi_1-\phi_2-\cdots-\phi_p}
\end{align}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

c = 2
phai = [0.1, -0.3, 1.1]

fig, ax = plt.subplots(1, 3, figsize=(16,4))
for i in range(3):
    y0 = 0
    y1 = [y0]
    for _ in range(100):
        y1_next = c + phai[i]*y1[-1] + np.random.normal(0, 1)
        y1.append(y1_next)
    ax[i].plot(y1)
```

### 定常過程と非定常過程
ある過程$y_t$を対象とし、平均や分散といった統計的な構造が時間と共にあまり変化せず、一定に落ち着いている場合、当該過程は定常的という。

> 定義1 弱定常性
時系列過程$y_t$に関し、その1次モーメント(期待値)と2次モーメント(分散、共分散)とが有限で不変であるとき、過程$y_t$は弱定常や共分散定常という。つまり、次の3つの条件を満たす過程$y_t$は弱定常や共分散定常という。
$$
E(y_t)=\mu<\infty\\
Var(y_t)=E\biggl[(y_t-\mu)^2 \biggr]=\sigma^2<\infty\\
Cov(y_t,y_{t-s})=E[(y_t-\mu)(y_{t-s}-\mu)]=\sigma(s),s>0
$$

これらの式は、平均・分散が不変であり、自己共分散が時点の差($s>0$)のみに依存し、時点$t$に依存していないことを表している。
弱定常性は1次モーメントと2次モーメントとに限定し、それらが時点$t$に依存していないことを求める。
強定常性は3次や4次モーメントを含めるが、普通の統計的議論で求められるのは、1次・2次モーメントであり、弱定常性で十分である。
「弱定常」は単に「定常」と呼ぶのが普通である。

定常過程と非定常過程の対比として次のAR(1)過程を考える。
$$
y_t=\rho y_{t-1}+u_t
$$
ただし、$u_t$はホワイトノイズとする。AR(1)の場合は「$係数\rho=固有値$」になる特殊ケースとなっている。
そこで固有値を$\rho$、そのスケールを$|\rho|$と表記する。$|\rho|=1$のスケールを持つ固有値$\rho$を単位根という。
スケール$|\rho|$の違いによって次のように分類できる。
1. $|\rho|<1$:定常過程
変数$u_t$に生じたショックの波及効果(限界効果)は時間の経過に伴って衰退していく。
2. $|\rho|=1$:非定常過程
変数$u_t$に生じたショックの波及効果(限界効果)は時間の経過に伴って衰退することなく一定で永遠に続く。
3. $|\rho|>1$:非定常過程
変数$u_t$に生じたショックの波及効果(限界効果)は時間の経過に伴って拡大し、爆発的な波及効果になる。(爆発的仮定)

ここで、係数$\rho$の値を(0.8,1.0,1.2)に設定した場合の図を示す。

```python
rhos = [0.8, 1, 1.2]

fig, ax = plt.subplots(1, 3, figsize=(16,4))
for i in range(3):
    y0 = 0
    y1 = [y0]
    for _ in range(100):
        y1_next = rhos[i]*y1[-1] + np.random.normal(0, 1)
        y1.append(y1_next)
    ax[i].plot(y1)
    ax[i].set_title('rho='+str(rhos[i]))
```

時期$h=0$において$\delta u_0=1$のショックが生じた想定にし、AR過程の波及効果(限界効果)$\Delta y_h(h=-1,0,1,\cdots)$を示している。

```python
lag = np.arange(10)
[plt.plot(lag, r**lag, label='rho='+str(r)) for r in rhos];
plt.legend();
```

定常性の定義が成立する領域は、いわゆる$|\rho|<1$の領域であり、大数の法則や中心極限定理など、標準的な漸近理論が通用する世界である。
一方、非定常の領域は、大数の法則や中心極限定理などが成立しない世界である。それは$|\rho|\geq1$の領域である。

### 和分過程
非定常過程の定常化は次の2つの概念に分類できる。
1. トレンド定常
2. 階差定常

これを受け、時系列過程の呼び名も次のように分ける。

1. トレンド定常過程：タイムトレンド(時間のみの関数)を取り除くことによって定常になる過程は、トレンド定常過程という。トレンド定常過程はレベル定常過程にタイムトレンドを加えた過程といえる。トレンド定常過程とレベル定常過程とは$|\rho|<1$といった共通性がある。
2. 階差定常過程：階差をとることによって定常となる過程は、階差定常過程という。差分定常過程とも呼ばれている。これらは$|\rho|\geq1$といった共通性がある。

階差定常過程は、何回階差を取れば定常過程になるかといった視点から分類される。
非定常過程$y_t$が1次の階差変換で定常となるとき、$y_t$は1次の和分過程と呼び、
$$
y_t\sim I(1)
$$
と表記する。
2次の和分過程$y_t$は、1回目の階差で$\Delta y_t\sim I(1)$となり、2回目の階差で$\Delta^2y_t\sim I(0)$となる。
原系列のままでも定常な過程やトレンド定常過程は$I(0)$過程に分類される。
和分過程の定義は次のように表現できる。
>定義 和分過程
$d-1$次の階差を取った過程は非定常であるが、$d$次の階差を取った過程が定常であるとき、過程は$d$次和分過程と呼び、$I(d)$過程と表記する。

```python
y1 = [0]
for _ in range(1000):
    y1_next = 0.8*y1[-1] + np.random.normal(0, 1)
    y1.append(y1_next)
y = np.array(y1).cumsum()
plt.plot(y);
```

### フィルター
次のAR(1)を考える。
$$
y_t=\rho y_{t-1}+u_t
$$
このAR(1)を次のように変換する。
$$
y_t-\rho y_{t-1}=+u_t\\
y_t-\rho Ly_t=+u_t\\
(1-\rho L)y_t=u_t\\
\rho(L)y_t=u_t
$$
ただし、$\rho(L)=1-\rho L$である。
このようなラグ多項式$\rho(L)$はフィルターと呼ばれる。
例えばフィルター$\rho(L)$を数$z$の多項式$\rho(z)$として表現し、$\rho(z)=0$と置けば、
$$
\rho(z)=1-\rho z=0
$$
といった方程式になる。
この方程式$\rho(z)=0$はAR過程$\rho(L)y_t$が定常か否かを判定する方程式として利用できる。
方程式$\rho(z)=0$を解く個々の根$z$が$|z|>1$であれば、当該AR過程$\rho(L)y_t$は単位根を持たず、定常である。
ここに示す具体例の場合、$\rho(z)=0$を解くと、$z=\frac{1}{\rho}$が得られる。
したがって、$|z|=\frac{1}{|\rho|}>1$である条件は$|\rho|<1$とも表現できる。
そして定常($|\rho|<1$)の場合は、$\rho(1)^{-1}$を次の幾何級数に変換できる。
$$
\begin{align}
\rho(1)^{-1}&=\frac{1}{1-\rho}\\
&=1+\rho+\rho^2+\cdots\\
&=\sum_{s=0}^\infty\rho^s
\end{align}
$$
ただし、$|\rho|<1$である。
この定理を利用し、次の幾何級数を得ることができる。
$$
\begin{align}
\rho(L)^{-1}&=\frac{1}{1-\rho L}\\
&=1+(\rho L)^1+(\rho L)^2+\cdots\\
&=\sum_{s=0}^\infty\rho^sL^s\\
\rho(L)^{-1}u_t&=\frac{1}{1-\rho L}u_t\\
&=u_t+\rho Lu_t+\rho^2 L^2u_t+\cdots\\
&=u_t+\rho u_{t-1}+\rho^2 u_{t-2}+\cdots\\
&=\sum_{s=0}^\infty\rho^su_{t-s}
\end{align}
$$
したがって、次の結果が得られる。
$$
\begin{align}
y_t&=\rho(L)^{-1}u_t\\
&=\sum_{s=0}^\infty\rho^su_{t-s}
\end{align}
$$
これは定常AR過程がMA過程として表現できることを表している。

次に示すような実数の係数$\psi_0,\psi_1,\cdots$を持つラグ多項式をフィルターという。
$$
\psi(L)=\psi_0+\psi_1L+\psi_2L^2\cdots
$$
または、
$$
\psi(L)=\psi_0+\psi_1L+\psi_2L^2\cdots+\psi_pL^p
$$
次に示す$p$次フィルターにおいて、多項式の根を求めるために$\psi(z)=0$と置くと、
$$
\psi(z)=\psi_0+\psi_1z+\psi_2z^2\cdots+\psi_pz^p=0
$$
といった方程式が得られる。ただし$z$は未知数としての数を表す。
方程式$\psi(z)=0$は、$\psi(L)$で示すデータ変換メカニズムが定常か否かを判別する方程式として利用できる。
ARMA過程では、次のフィルターを使用して表現していた。
$$
\phi(L)=1-\phi_1L-\phi_2L^2-\cdots-\phi_pL^p\\
\theta(L)=1+\theta_1L+\theta_2L^2+\cdots+\theta_qL^q\\
$$
したがって、$\phi(z)=0$も$\theta(z)=0$も、それらのデータ変換メカニズムが定常か否かを判別する方程式として利用できる。

```python
rho = 0.1

y0 = np.sum([(rho**s)*np.random.normal(0, 1) for s in range(2000)])
y1 = [y0]
for i in range(100):
    y0 = np.random.normal(0, 1) + rho*y0
    y1.append(y0)
plt.plot(y1)
```

```python
y0 = 0
y1 = [y0]
for _ in range(100):
    y1_next = rho*y1[-1] + np.random.normal(0, 1)
    y1.append(y1_next)
plt.plot(y1)
```

### AR過程の定常条件
ここでは次のAR(p)過程を対象とし、定常性を得るための条件について考える。
$$
y_t=\phi_1y_{t-1}+\phi_2y_{t-2}+\cdots+\phi_py_{t-p}+c+u_t\\
u_t\sim WN(0,\sigma^2)
$$
差分方程式を解く手順に従い、ラグ方程式を方程式左辺に移動させる。
$$
y_t-\phi_1y_{t-1}-\phi_2y_{t-2}-\cdots-\phi_py_{t-p}=c+u_t
$$
これは非同次方程式と呼ばれる形式になっている。右辺は定常である。
一方、同次方程式は次の形式になる。
$$
y_t-\phi_1y_{t-1}-\phi_2y_{t-2}-\cdots-\phi_py_{t-p}=0
$$
AR(p)をAR(1)形式に変換することにより、次のAR(1)過程システムが得られる。
$$
\boldsymbol{y}^*_t=\boldsymbol{\phi}^*\boldsymbol{y}^*_{t-1}+\boldsymbol{c}^*+\boldsymbol{u}^*_t
$$
この内部構造は次のようになっている。
$$
\left(\begin{array}{c}
y\\
y_{t-1}\\
y_{t-2}\\
\vdots\\
y_{t-p+1}
\end{array}
\right)=
\left(\begin{array}{cccc}
\phi_1&\phi_2&\phi_3&\cdots&\phi_p\\
1&0&0&\cdots&0\\
0&1&0&\cdots&0\\
\vdots&\vdots&\vdots&\vdots&\vdots\\
0&0&0&1&0\\
\end{array}
\right)
\left(\begin{array}{c}
y_{t-1}\\
y_{t-2}\\
y_{t-3}\\
\vdots\\
y_{t-p}
\end{array}
\right)+
\left(\begin{array}{c}
c\\
0\\
0\\
\vdots\\
0
\end{array}
\right)+
\left(\begin{array}{c}
u_t\\
0\\
0\\
\vdots\\
0
\end{array}
\right)
$$
AR(p)過程が定常か否かは、$p$個の係数$(\phi_1,\phi_2,\cdots,\phi_p)$からなるデータ変換メカニズムに関わっている。
$p×p$の行列$\boldsymbol{\phi}^*$の特性に関わっているともいえる。数学的には固有値の話となっている。
固有値を求めるために次の固有方程式を用いる。
$$
det(\boldsymbol{\phi}^*-\lambda\boldsymbol{I})=0
$$
分解すると、
$$
\lambda^{p}-\phi_1\lambda^{p-1}-\phi_2\lambda^{p-2}-\cdots-\phi_{p-1}\lambda-\phi_p=0
$$
$p=1$の場合、固有方程式の根$\lambda$は係数$\phi_1$に等しい。
$|\lambda|<1$であれば、AR(1)過程は定常である。それ以外は非定常となる。

AR(p)家庭の場合においては、固有方程式は$\lambda$の$p$次方程式であるから、それを解く根が$p$個存在し、それらを$\{\lambda_1,\lambda_2,\cdots,\lambda_p\}$とする。
$$
|\lambda_i|<1
$$
であれば、当該AR(p)過程は定常である。つまりI(0)過程である。
$$
|\lambda_i|\geq1
$$
を満たす根が1個でもあれば、当該AR(p)過程は非定常である。具体的にはI(1)以上の和分過程である。

```python
rho = np.array([0.1, 0.5])

y0 = 0
y1 = [y0, y0]
for i in range(200):
    y0 = np.random.normal(0, 1) + rho[0]*y1[-1] + rho[1]*y1[-2]
    y1.append(y0)
plt.plot(y1)
```

```python
Phi = np.concatenate([rho.reshape(1,-1), np.eye(len(rho))], axis=0)[:-1]
eig_values, eig_vectors = np.linalg.eig(Phi)

print('eigen values: ', eig_values)
np.all(eig_values<1)
```

### 固有方程式とラグ型特性方程式
AR(p)過程の定常性を判別するための方程式は、固有方程式と共に、次の方程式も利用されている。
$$
\phi(z)=1-\phi_1z-\phi_2z^2-\cdots-\phi_pz^p=0
$$
ラグ演算子$L$を$z$に置き換えた特性方程式といった意味で、こちらはラグ特性方程式と呼ぶ。
ラグ特性方程式を用いる場合は、個々の根が$|z_i|>1$であれば、当該AR(p)過程は定常である。
それ以外、つまり$|z_i|\leq1$の根が1つでも存在すれば非定常である。
$\lambda$と$z$の関係は次のようになる。
$$
\lambda=\frac{1}{z}
$$

### 定常なAR(1)過程の期待値と分散
次に示す定常なAR(1)過程を用い、$y_t$の期待値や分散を計算する道筋を示す。
$$
y_t=\rho y_{t-1}+c+u_t,\hspace{3mm}y_t\sim WN(0,\sigma^2)
$$

ただし、$c$は定数項、$\rho$は定常となるための性質$|\rho|<1$をもつ係数である。
定常であることから$E(y_t)=E(y_{t-1})=\mu$と表記する。
$$
\begin{align}
E(y_t)&=E(\rho y_{t-1}+c+u_t)\\
&=\rho E(y_{t-1})+c+0\\
\mu&=\rho\mu+c
\end{align}
$$
最終的に期待値$E(y_t)$は次のように求まる。
$$
E(y_t)=\mu=\frac{c}{1-\rho}
$$

次に分散$Var(y_t)$を求める。
AR(1)過程を次のように変換する。
$$
(1-\rho L)y_t=c+u_t\\
\rho(L)y_t=c+u_t
$$
ただし、$\rho(L)=1-\rho L$である。$|\rho|<1$であるから次の反転が可能である。
$$
y_t=\rho(L)^{-1}(c+u_t)
$$
そして次の結果を得ることができる。
$$
\begin{align}
\rho(L)^{-1}c&=\frac{c}{1-\rho L}\\
&=s\sum_{s=0}^\infty(\rho L)^s\\
&=s\sum_{s=0}^\infty(\rho)^s\\
&=\frac{c}{1-\rho}\\
&=\mu\\
\rho(L)^{-1}u_t&=\frac{1}{1-\rho L}u_t\\
&=u_t+\rho Lu_t+\rho^2L^2u_t+\cdots\\
&=u_t+\rho u_{t-1}+\rho^2u_{t-2}+\cdots\\
&=\sum_{s=0}^\infty \rho^su_{t-s}
\end{align}
$$
したがって、定常AR(1)過程は次のようにMA過程として表現できる。
$$
\begin{align}
y_t&=\rho(L)^{-1}(c+u_t)\\
&=\mu+\sum_{s=0}^\infty\rho^su_{t-s}
\end{align}
$$
そして分散は次のように求める。
$$
\begin{align}
Var(y_t)&=Var\biggl(\mu+\sum_{s=0}^\infty\rho^su_{t-s} \biggr)\\
&=0+\sum_{s=0}^\infty\rho^{2s}Var(u_{t-s})\\
&=\sum_{s=0}^\infty\rho^{2s}\sigma^2\\
&=\frac{\sigma^2}{1-\rho^2}
\end{align}
$$

次の情報が重要である。
1. 定常なAR過程：分散が有限である。幾何級数の定理を使って分散を求めることができる。
2. 非定常なAR過程：分散が有限でない。幾何級数の定理を使って分散を求めることができない。

```python
rho = 0.3
c = 2
sig = 1.2

y0 = 0
y1 = [y0, y0]
for i in range(300):
    y0 = c + np.random.normal(0, sig) + rho*y0
    y1.append(y0)

fig, ax = plt.subplots(1, 2, figsize=(12,4))
ax[0].plot(y1);
ax[1].hist(y1, bins=25);
```

```python
print('E(y): ', np.mean(y1))
print('c/(1-rho): ', c/(1-rho))

print('\nVar(y): ', np.var(y1))
print('sigam^2/(1-rho^2): ', sig**2/(1-rho**2))
```

### 定常AR過程のMA表現
AR(1)過程の自己回帰構造を利用して式の挿入を繰り返すと、次の関係式になってくる。
$$
\begin{align}
y_t&=\rho y_{t-1}c+u_t\\
&=\rho(\rho y_{t-2}c+u_{t-1})+c+u_t\\
&=\rho^2y_{t-2}+c(1+\rho)+u_t+\rho u_{t-1}\\
&=\rho^2(\rho y_{t-3}+c+u_t)+c(1+\rho)+u_t+\rho u_{t-1}\\
&=rho^3y_{t-3}+c(1+\rho+\rho^2)+u_t+\rho u_{t-1}+\rho^3u_{t-2}\\
&=\rho^hy_{t-h}+c\sum_{s=0}^{h-1}\rho^s+\sum_{s=0}^{h-1}\rho^su_{t-s}
\end{align}
$$
$|\rho|<1$であるので、$h→\infty$のとき$\rho^hy_{t-h}$はゼロに確率収束し、次の関係が成立する。
$$
y_t=c\sum_{s=0}^{h-1}\rho^s+\sum_{s=0}^{h-1}\rho^su_{t-s}
$$
幾何級数の定理を使用して、次の関係式を得る。
$$
\begin{align}
y_t&=\frac{c}{1-\rho}+\sum_{s=0}^{h-1}\rho^su_{t-s}\\
&=\mu+\sum_{s=0}^{h-1}\rho^su_{t-s}\\
u_t&\sim WN(0,\rho^2)
\end{align}
$$
ただし、
$$
\mu=\frac{c}{1-\rho}\\
\sum_{s=0}^\infty|\rho^s|<\infty
$$
である。これはMA過程であり、MA($\infty$)と表す。
AR過程は、自己回帰構造を利用して式の挿入を繰り返すことができる乗数構造になっている。
このことが理論的な解析を難しくするが、この問題をなくしたのがMA($\infty$)表現である。
MA($\infty$)表現を利用すれば、

1. 定常AR過程の長期予測は過去の値・初期値$y_0$に依存しない
2. 観測値と期待値のギャップ($y_t-\mu$)は、撹乱項($u_t,u_{t-1},\cdots$)の加重和によって説明できる
3. その加重和の重み$\rho^s$は指数的に減衰していくことが明確に分かる。これは現在から遠い時点の$u_{t-s}$ほど、現在の$y_t$への影響が小さいことを表している。いわゆる過渡的な波及効果になる。

MA($\infty$)表現における($u_t,u_{t-1},\cdots$)はイノベーションと呼ばれる。

```python
c = 2
rho = 0.1

y0 = np.sum([(rho**s)*np.random.normal(0, 1) for s in range(2000)])
y1 = [y0]
for i in range(1000):
    y0 = c+np.random.normal(0, 1) + rho*y0 # \rho×(u_t-1+rho ut-2+rho^2 ut-3+) 結局rho×y0となる
    y1.append(y0)
plt.plot(y1)
```

### MA過程の反転可能性
MA過程も、ある一定の条件を満たすと、AR過程として表現できる。
まず、次の一般的なAR(p)を考える。
$$
\begin{align}
y_t&=\phi_1y_{t-1}+\phi_2y_{t-2}+\cdots+\phi_py_{t-p}+c+u_t\\
u_t&\sim WN(0,\sigma^2)
\end{align}
$$
フィルターを使うと、AR(p)過程は次のようにコンパクトに表現できる。
$$
\phi(L)y_t=c+u_t
$$
ただし、
$$
\phi(L)=1-\phi_1L^1-\phi_2L^2-\cdots-\phi_pL^p
$$
である。
定常なAR(p)過程は次のようにMA過程として書ける。
$$
y_t=\mu+\psi(L)u_t,\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$
ただし、
$$
\begin{align}
\psi(L)&=\phi(L)^{-1}\\
&=\frac{1}{1-\phi_1L^1-\phi_2L^2-\cdots-\phi_pL^p}\\
&=1+\psi_1L^1+\psi_2L^2+\cdots\\
\mu&=\phi(L)^{-1}c\\
&=\phi(1)^{-1}c\\
&=\frac{c}{1-\phi_1-\phi_2-\cdots-\phi_p}\\
&\sum_{s=0}^\infty |\psi_s|<\infty,\hspace{2mm}\psi_0=1
\end{align}
$$
である。

次のMA(1)過程を考える。
$$
y_t=\mu+u_t+\theta u_{t-1},\hspace{3mm}u_t\sim WN((0,\sigma^2)
$$
この場合は$\theta(L)=1+\theta L$をフィルターとし、次のラグ型特性方程式を考える。
$$
\theta (z)=1;\theta z=0
$$
このラグ特性方程式の解は$z=-\frac{1}{\theta}$である。
したがって、$|z|=\frac{1}{|\theta|}>1$となる条件は$|\theta|<1$である。
$|\theta|<1$であれば、次の幾何級数が成立する。
$$
\begin{align}
\frac{1}{1+\theta}&=\frac{1}{1-(-\theta)}\\
&=\sum_{s=0}^\infty(-\theta)^s\\
&=1-\theta+\theta^2-\theta^3+\cdots\\
\theta(L)^{-1}&=\frac{1}{1+\theta L}\\
&=\frac{1}{1+-(-\theta L)}\\
&=\sum_{s=0}^\infty(-\theta L)^s\\
&=1-\theta L+\theta^2L^2-\theta^3L^3+\cdots
\end{align}
$$
そしてMA(1)過程は次のように反転可能である。
$$
y_t-\mu=\theta(L)u_t\\
\theta(L)^{-1}(y_t-\mu)=u_t
$$
さらに次の結果を得ることができる。
$$
\begin{align}
\theta(L)^{-1}\mu&=\frac{1}{1+\theta L}\mu\\
&=\frac{\mu}{1+\theta}\\
\theta(L)^{-1}y_t&=\frac{1}{1+\theta L}y_t\\
&=\sum_{s=0}^\infty (-\theta L)^sy_t\\
&=(1-\theta L+\theta^2 L^2-\theta^3 L^3+\cdots)y_t\\
&=y_t-\theta y_{t-1}+\theta^2 y_{t-2}-\theta^3 y_{t-3}+\cdots
\end{align}
$$
最終的には次のAR($\infty$)表現が得られる。
$$
\phi(L)y_t=c+u_t
$$
ただし、
$$
\begin{align}
\phi(L)y_t&=\theta(L)^{-1}y_t\\
&=\frac{1}{1+\theta L}y_t\\
&=y_t-\theta y_{t-1}+\theta^2 y_{t-2}-\theta^3 y_{t-3}+\cdots\\
c&=\theta(L)^{-1}\mu\\
&=\theta(1)^{-1}\mu\\
&=\frac{\mu}{1+\theta}\\
&\sum_{s=0}^\infty|\theta^s|<\infty
\end{align}
$$



```python
theta = 0.5
s = 1.2
mu = 1

y1 = []
u = [np.random.normal(0, s)]
for _ in range(300):
    u.append(np.random.normal(0, s))
    y1_next = mu + u[-1] + theta*u[-2]
    y1.append(y1_next)
plt.plot(y1)
```

```python
theta = 0.5
s = 1.2
mu = 1
c = mu/(1+theta)

y1 = [1]
u = []
for _ in range(300):
    u.append(np.random.normal(0, s))
    y1_next = c + u[-1] + sum([-(-theta)**i * y1[-i] for i in range(1,len(y1))])
    y1.append(y1_next)
plt.plot(y1)
```

次のMA(q)を考える。
$$
y_t=\mu+\theta(L)u_t,\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$
ただし、
$$
\theta (L)=1+\theta_1L^1+\theta_2L^2+\theta_3L^3+\cdots+\theta_qL^q
$$
である。
ラグ特性方程式$\theta (z)=0$を満たす個々の$z_i$が$|z_i|>1$であれば、MA(q)はAR($\infty$)へ反転可能である。
$$
\phi(L)y_t=c+u_t,\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$
ただし、
$$
\begin{align}
\phi(L)&=\theta(L)^{-1}\\
&=\frac{1}{1+\theta_1L^1+\theta_2L^2+\theta_3L^3+\cdots+\theta_qL^q}\\
&=1-\phi_1L^1-\phi_2L^2-\cdots\\
c&=\theta(L)^{-1}\mu\\
&=\theta(1)^{-1}\mu\\
&=\frac{1}{1+\theta_1+\theta_2+\cdots+\theta_q}\\
&\sum_{s=0}^\infty|\phi_s|<\infty,\hspace{2mm}\phi_0=1
\end{align}
$$
である。

### ARMA過程の定常条件
次のARMA過程を考え、定常性の条件を考える。
$$
\phi(L)y_t=c+\theta(L)u_t,\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$
または、
$$
\phi(L)(y_t-\mu)=\theta(L)u_t,\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$
ただし、
$$
\begin{align}
\phi(L)&=1-\phi_1L^1-\phi_2L^2-\cdots-\phi_pL^p\\
\theta(L)&=1+\theta_1L^1+\theta_2L^2+\cdots+\theta_qL^q\\
\mu&=\phi(L)^{-1}\\
&=\phi(1)^{-1}\\
&=\frac{c}{1-\phi_1-\phi_1-\cdots-\phi_p}
\end{align}
$$
である。
MA過程は常に定常であることから、ARMA過程が定常か否かは、そのAR過程に依存する。
したがって、AR過程に関する次のラグ特性方程式$\phi(z)=0$を用いて判別することができる。
$$
\phi(z)=1-\phi_1z-\phi_2z^2-\cdots-\phi_pz^p=0
$$
ラグ特性方程式$\phi(z)=0$を満たす個々の根$z_i$が$|z_i|>1$であれば、当該ARMA過程は定常である。
定常なAR過程はMA過程として表現できた。したがって、定常なARMA過程もMA過程として表現でき、次のMA表現を持つ。
$$
y_t=\mu+\psi(L)u_t
$$
ただし、
$$
\begin{align}
\psi(L)&=\phi(L)^{-1}\theta(L)\\
&=\frac{1+\theta_1L^1+\theta_2L^2+\cdots+\theta_qL^q}{1-\phi_1L^1-\phi_2L^2-\cdots-\phi_pL^p}\\
&=1+\psi_1L^1+\psi_2L^2+\cdots\\
\mu&=\phi(L)^{-1}c\\
&=\phi(1)^{-1}c\\
&=\frac{c}{1-\phi_1-\phi_2-\cdots-\phi_p}\\
&\sum_{s=0}^\infty|\psi_s|<\infty,\ \psi_0=1
\end{align}
$$
である。

```python
theta = np.array([0.1, 0.5])
phi = np.array([1.6])

s = 1.0
c = 2

y0 = 0
y1 = [y0, y0]
u = []
for i in range(200):
    u.append(np.random.normal(0, s))
    y0 = np.random.normal(0, 1) + theta[0]*y1[-1] + theta[1]*y1[-2] + phi*u[-1]
    y1.append(y0)
plt.plot(y1)
```

### ARIMAモデル
階差をとった定常過程をARMA過程として定式化するモデルは、ARIMAモデルと呼ばれ、$ARIMA(p,d,q)$と表記する。
ただし、$d$は和分の次数、$p$は自己回帰の次数、$q$は移動平均の次数を表す。
$ARIMA(p,d,q)$は次のように定式化できる。
$$
(1-\phi_1L^1-\phi_2L^2-\cdots-\phi_pL^p)(1-L)^dy_t=(1+\theta_1L^1+\theta_2L^2+\cdots+\theta_qL^q)u_t
$$
フィルターを使って次のようにコンパクトに表現することもできる。
$$
\phi(L)(1-L)^dy_t=\theta(L)u_t
$$
ただし、
$$
\phi(L)=1-\phi_1L^1-\phi_2L^2-\cdots-\phi_pL^p\\
\theta(L)=1+\theta_1L^1+\theta_2L^2+\cdots+\theta_qL^q
$$
である。
ある時系列データが$ARIMA(p,d,q)$のどのタイプに属するかを同定する道具としては自己相関関数と偏自己相関関数がある。
自己相関関数は時間差$s$を置いた$y_t$と$y_{t-s}$との相関を計測する。
期間$s$に存在する中間の$y_{t-i}(i=1,2,\cdots,s-1)$の影響を制御した状態の、$y_t$と$y_{t-s}$との相関を計測するのが偏自己相関関数である。

自己相関と偏自己相関の性質は次のようになる。

|モデル|自己相関|偏自己相関|
|-|-|-|
|AR(p)|次第に減衰|ラグ次数p+1以降はゼロ|
|MA(q)|ラグ次数q+1以降はゼロ|次第に減衰|
|ARMA(p,q)|次第に減衰|次第に減衰|

このパターンからモデルとラグ次数を同定していく。
ただし、自己相関と偏自己相関だけでは$ARMA(p,q)$の次数は決定できない。

### I(0)過程の性質とI(1)過程の性質
I(0)過程にはレベル定常過程とトレンド定常過程が含まれている。
時系列過程の性質を論じる場面における「定常過程」対「単位根過程」は、内実は「I(0)過程」対「I(1)過程」である。

定常なARMA過程は次のMA過程として表現できる。
$$
y_t=\mu+\psi(L)u_t\\
u_t\sim WN(0,\sigma^2)
$$

ただし、

$$
\psi(L)=1+1+\psi_1L^1+\psi_2L^2+\psi_3L^3+\cdots\\
\sum_{s=0}^\infty |\psi_s|<\infty,\hspace{3mm}\psi_0=1
$$
である。

トレンド定常過程はタイムトレンドを除去すると定常となる過程であるから、それは次のようにモデル化できる。
$$
y_t=c_1+c_2t+\psi(L)u_t,\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$

ここで、$c_2t$はタイムトレンド項という。
$I(0)$過程の性質はMA過程$\psi(L)u_t$を単純化した次のモデルを使って議論した方が便利である。
$$
y_t=c_1+c_2t+u_t,\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$

このモデルを使って$I(0)$過程の性質を調べる。

$y_t$から$c_1+c_2t$を取り除いた場合の期待値と分散を調べる。
$$
E(y_t-c_1-c_2t)=E(u_t)=0\\
Var(y_t-c_1-c_2t)=Var(u_t)=\sigma^2
$$
トレンド定常過程からトレンド成分を取り除いた残差$y_t-c_1-c_2t$は、定常過程の条件を満たしている。
このようにトレンド定常過程は、タイムトレンドの周りで定常である。
一方、I(1)過程は1次の階差で定常となる過程であるから、
$$
\Delta y_t=c+\psi(L) u_t,\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$
と置き、次の形へ変換することでモデル化できる。
$$
y_t=y_{t-1}+c+\psi(L)u_t,\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$
ただし、$\psi(1)\neq1$とする。
$I(1)$過程の性質も、MA過程$\psi(L)u_t$を単純化した次のモデルを使って議論した方が便利である。
$$
y_t=y_{t-1}+c+u_t,\hspace{3mm}u_t\sim WN(0,\sigma^2)
$$
定数項$c$はドリフト項と呼ばれる。ドリフト付きI(1)過程ということもある。
ドリフト付きI(1)過程はタイムトレンドが表現できる。
$$
\begin{align}
y_t&=c+y_{t-1}+u_t\\
&=c+(c+y_{t-2}+u_{t-1})+u_t\\
&=2c+y_{t-2}+(u_t+u_{t-1})\\
&=2c+(c+y_{t-3}+u_{t-2})+(u_t+u_{t-1})\\
&\vdots\\
&=ct+y_0+\sum_{s=0}^{t-1}u_{t-s}\\
&=ct+y_0+v_t
\end{align}
$$

ただし、$v_t=u_t+u_{t-1}+\cdots+u_1$である。
$ct$は確定的トレンドと呼ばれ、$v_t=\sum_{s=0}^{t-1}u_{t-s}$は確率的トレンドと呼ばれる。
$v_t=\sum_{s=0}^{t-1}u_{t-s}$に示すように、確率的トレンドは波及効果が減衰しないショックの和から成り立っている。
つまり、一度生じたショックは恒久的な影響を将来にわたって持つ。
I(1)過程はタイムトレンドを取り除いても定常にならない性質がある。
$$
\begin{align}
E(y_t-ct)&=E(y_0+v_t)\\
&=y_0+E(u_t+u_{t-1}+\cdots+u_1)\\
&=y_0\\
Var(y_t-ct)&=Var(y_0+v_t)\\
&=0+Var(u_t+u_{t-1}+\cdots+u_1)\\
&=\sigma^2t
\end{align}
$$

```python
c1 = 1
c2 = 0.05
sigma2 = 1.5

y2_sum = np.zeros(200).reshape(1,200)
fig, ax = plt.subplots(1,2, figsize=(12,4))
for k in range(300):
    y0 = 0
    y1 = np.array([y0])
    y2 = np.array([y0])
    for i in range(199):
        y0 = c1 + c2*i + np.random.normal(0, np.sqrt(sigma2))
        y1 = np.append(y1, y0)
        y2 = np.append(y2, y0 -c1 - c2*i)
    y2_sum = np.concatenate([y2_sum, y2.reshape(1,200)], axis=0)

ax[0].plot(y1)
ax[1].plot(y2)
```

```python
y_mean = np.mean(y2_sum[1:, :], axis=0)
y_var = np.var(y2_sum[1:, :], axis=0)
plt.plot(y_mean)
plt.plot(y_var)
```

### I(2)過程の性質
階差は離散型の微分に相当し、I(0)の累積効果はI(1)の限界効果になり、I(1)の累積効果はI(2)の限界効果になる関係がある。
したがって、次の推論が成り立つ。

|過程のタイプ|限界効果|累積効果|
|-|-|-|
|I(0)|ゼロに収束|一定になる|
|I(1)|時間に対して一定|時間に対して線形関数的|
|I(2)|時間に対して線形関数的|時間に対して2次関数的|

$\Delta^2 y_t=c+u_t$とおくと、次のI(2)基準過程モデルを作ることができる。

$$
y_t=2y_{t-1}-y_{t-2}+c+u_t
$$

ただし、$c$は定数項、$u_t$はホワイトノイズである。
I(2)過程において、$\Delta u_0=1$のショックを発生させ、当該$I(2)$基準過程がどのような限界効果と累積効果を表すかを調べる。

```python
c= 0.5
rho1 = 2
rho2 = -1
sigma2 = 1.5

fig, ax = plt.subplots(1,2, figsize=(12,4))
y0 = 0
y1 = np.array([y0]*2)
y2 = np.array([y0])
for i in range(300):
    if i == 0:
        y0 = c + rho1*y1[-1] + rho2*y1[-2] + 1
    else: 
        y0 = c + rho1*y1[-1] + rho2*y1[-2]#+ np.random.normal(0, np.sqrt(sigma2))
    y1 = np.append(y1, y0)
    y2 = np.append(y2, y0 -y1[-2])

ax[0].plot(y1)
ax[1].plot(y2)
```

ある時点で生じた1つのショックのI(2)基準過程への限界効果は線形関数的に拡大し、累積効果は2次関数的に拡大することが確認できた。

I(2)基準過程のための固有方程式は次のようになる。
$$
\lambda^2-2\lambda+1=0
$$
この方程式を解く根は次のようになる。
$$
\lambda=\frac{2±\sqrt{4-4}}{2}=1
$$
多重根となっているので、この固有方程式は次のように表現するのが適切である。
$$
(\lambda-1)^2=0
$$
I(2)基準過程は重複度2の単位根を持っている。

### 爆発的な現象
非定常な領域$|\rho|>1$では次のような基本的な問題がある。
標準的な漸近理論が成立しない領域であり、標準的な推定法では推定結果が漸近的に有限にならない。
したがって、爆発的な現象は回帰モデルで表現したほうが良い。

I(2)基準過程モデルは次のように書ける。
$$
y_t=2y_{t-1}-y_{t-2}+c+u_t
$$
このモデルは、固有方程式が重複度2の単位根$\lambda=1$を持つときの条件から構築している。
$$
(\lambda-1)^2=\lambda^2-2\lambda+1=0
$$
I(2)基準過程モデルは爆発的な現象に適している。
さらに爆発的な現象を表現したい場合は、次の$I(3)$基準過程モデルが考えられる。
$$
y_t=3y_{t-1}-3y_{t-2}+y_{t-1}+c+u_t
$$
I(3)基準過程モデルは、固有方程式が重複度2の単位根$\lambda=1$を持つときの条件から構築している。
$$
(\lambda-1)^3=\lambda^3-3\lambda+3\lambda-1=0
$$
I(3)基準過程モデルは、限界効果が2次関数的に拡大し、累積効果が3次関数的に拡大する。
このように単位根の乗数構造$(\lambda-1)^d$を利用すれば、さまざまな爆発的現象を記述する$I(d)$過程モデルをつくることができる。

```python
c= 0.5
rho1 = 3
rho2 = -3
rho3 = 1
sigma2 = 1.5

fig, ax = plt.subplots(1,2, figsize=(12,4))
y0 = 0
y1 = np.array([y0]*3)
y2 = np.array([y0])
for i in range(300):
    if i == 0:
        y0 = c + rho1*y1[-1] + rho2*y1[-2] + rho3*y1[-3] + 1
    else: 
        y0 = c + rho1*y1[-1] + rho2*y1[-2] + rho3*y1[-3]#+ np.random.normal(0, np.sqrt(sigma2))
    y1 = np.append(y1, y0)
    y2 = np.append(y2, y0 -y1[-2])

ax[0].plot(y1)
ax[1].plot(y2)
```

### さまざまな和分過程の限界効果と累積効果
I(d)基準過程に関する限界効果と累積効果については、次のようにまとめることができる。

|和分過程|限界効果|累積効果|
|-|-|-|
|I(0)基準過程|ショック発生時を除きゼロ|一定のレベルになる|
|I(1)基準過程|時間に対して一定になる|時間に対して線形関数になる|
|I(2)基準過程|時間に対して線形関数になる|時間に対して2次関数になる|
|I(3)基準過程|時間に対して2次関数になる|時間に対して3次関数になる|

和分過程の2分割化の視点を示す。

1. 任意なI(d)過程=和分過程+残りの定常部分
2. 任意なI(d)過程=推定しない部分+推定する部分
3. 任意なI(d)過程=I(d)基準過程+残りのI(0)過程
4. 任意なI(d)過程=衰退しない波及効果の部分+衰退する波及効果の部分

上に示す2分割化を用い、I(d)基準効果の性質と普通のI(0)過程の性質とを組み合わせれば、さまざまなI(d)過程の性質が想像でき、説明することができる。

### 固有方程式と固有値の解釈
固有方程式がARシステムの総合影響力を計測する計測器であることを示す。
個々の係数が$y_t$への影響力を表す数値であるから、それらの数値から構成されるベクトルや行列の固有値が$y_t$1への影響力に関係していることは明らかである。
AR(1)の場合はARシステム全体の影響力が11個の固有値に現れる特殊性がある。
AR(1)システムの影響力が固有値に等しいことは、
$$
\rho=\lambda
$$
と表記していた。
ラグ変数が複数あり、固有値が複数ある場合は、それら固有値全体のスケールがARシステム全体の影響力を表すと推察できる。
「他の事情同一の条件」のもとで、複数の係数の総合影響力を測る。
例として次のAR(3)過程を考える。
$$
y_t=\phi_1y_{t-1}+\phi_2y_{t-2}+\phi_3y_{t-3}+c+u_t\\
u_t\sim WN(0,\sigma^2)
$$
3つの係数$(\phi_1,\phi_2,\phi_3)$の総合影響力を$y_t$の影響力1から取り除いたときにゼロとなるように設定すれば、係数$(\phi_1,\phi_2,\phi_3)$の総合影響力が計測できる。
$$
y_t-\phi_1y_{t-1}-\phi_2y_{t-2}-\phi_3y_{t-3}=c+u_t
$$
そして次のように置く。
$$
y_t-\phi_1y_{t-1}-\phi_2y_{t-2}-\phi_3y_{t-3}=0
$$

この関係式は4個の係数$(1,-\phi_1,-\phi_2,-\phi_3)$を含み、各係数は異時点における影響力を表している。
同一の時点の影響力に変換する必要があるが、階差が乗数型の変換システムであることから、2期の差異は2乗となり、乗数の形で時間の差異を制御する。
つまり、2基の差異は$\lambda^0=1$対$\lambda^2$といった形で制御する。
4個の係数の影響力が正味でゼロとなる条件は、次の表現となる。
$$
\lambda^3-\phi_1\lambda^2-\phi_2\lambda^1-\phi_3=0
$$

これが固有方程式(特性方程式)である。
固有方程式の根(固有値)は影響力の「現在価値」に相当する数値である。

この固有方程式を解く根は3個あり、$\{\lambda_1,\lambda_2,\lambda_3\}$と表記する。
各根が次の条件を満たすとき、当該AR(3)過程は定常、つまりI(0)過程である。
$$
|\lambda_i|<1
$$
この条件は、個々のラグ変数が$y_t$自身の影響力1よりも小さいことを指している。
一方、次の条件を満たす根が1個でも存在するとき、当該AR(3)過程は非定常であり、I(1)以上の過程になる。
$$
|\lambda_i|\geq1
$$
この条件は、$y_t$自身の影響力1よりも大きいか等しい影響力を持つラグが存在していることを表している。
次の評価システムを作ることができる。
$$
1-\phi_1z-\phi_2z^2-\phi_3z^3=0
$$
こちらはラグ型特性方程式と呼ぶ。

### 自然対数に関する基本情報
自然対数$\log Y$を用いる理由は次の通りである。

1. $Y$よりも$\log Y$の方が定常性を得やすい
2. $Y$よりも$\log Y$の方が均一分散を得やすい
3. $Y$から$\log Y$へ変換しても、時系列の特性に重大な差異が生じない
4. $\log Y$の差分は変化率$\frac{\Delta Y}{Y}$を表し、それは汎用性の広い測度となる

利子率や失業率のように既に割合の形になっている変数は、自然対数にする理由がなくそのままの形で使うのが一般的である。
自然対数の差分は、次に示すように変化率を表す。
$$
\Delta \log Y=\frac{\Delta Y}{Y}
$$

この関係式は厳密には近似式であり、数%ぐらいの変化率であれば正確である。
変換率が大きい場合には、次の関係式で正確に計算できる。
$$
\frac{\Delta \log Y}{Y}=\exp(b)-1
$$

例えば、$\Delta \log Y=0.25$の場合で比較すると、次のようになる。

1. 近似式の場合：$\Delta \log Y=\frac{\Delta Y}{Y}=0.25$となり、25％の増加となる。
2. 正確な式の場合：$\Delta \log Y=\exp(0.25)-1=0.284$となり、28.4%の増加となる。

回帰分析における具体例を考える。次の推定された回帰モデルを考える。

$$
\log Y=1.23+0.45\log X
$$

勾配係数の推定値0.45は、$X$に対する$Y$の弾力性を表す。つまり、
$$
\frac{\Delta Y}{\Delta X}=\frac{\frac{\Delta Y}{Y}}{\frac{\Delta X}{X}}=0.45=\frac{0.45%}{1%}
$$
を表す。
したがって説明変数$X$が1%増加すると、従属変数$Y$は平均で0.45%増える傾向があるといった表現になる。
