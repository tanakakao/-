#### 実フーリエ級数を見直す
オイラーの公式
$$
e^{i\theta}=\cos\theta+i\sin\theta
$$
から、
$$
\cos\theta=\frac{e^{i\theta}+e^{-i\theta}}{2}\\
\sin\theta=\frac{e^{i\theta}-e^{-i\theta}}{2i}
$$
となることがわかる。
基本周波数$f_0=1/T$とすると、
$$
\begin{align}
x(t)&\sim \frac{a_0}{2}+\sum_{n=1}^\infty(a_n\cos2\pi f_0nt+b_n\sin2\pi f_0nt)\\
&=\frac{a_0}{2}+\sum_{n=1}^\infty\biggl(a_n\frac{e^{2\pi if_0nt}+e^{-2\pi if_0nt}}{2}+b_n\frac{e^{2\pi if_0nt}-e^{-2\pi if_0nt}}{2i} \biggr)\\
&=\frac{a_0}{2}+\sum_{n=1}^\infty\biggl(a_n\frac{e^{2\pi if_0nt}+e^{-2\pi if_0nt}}{2}+b_n\frac{-ie^{2\pi if_0nt}+ie^{-2\pi if_0nt}}{2i} \biggr)\\
&=\frac{a_0}{2}+\sum_{n=1}^\infty\biggl(\frac{a_n-ib_n}{2}e^{2\pi if_0nt}+\frac{a_n+ib_n}{2}e^{-2\pi if_0nt} \biggr)\\
&=\sum_{n=-\infty}^{\infty}c_ne^{2\pi if_0nt}
\end{align}
$$
ここで、
$$
c_0=\frac{a_0}{2},\ c_n=\frac{a_n-ib_n}{2},\ c_{-n}=\bar{c_n}=\frac{a_n+ib_n}{2}
$$
とおいた。
フーリエ係数を積分の形で表すと次のようになる。
$$
c_n=\frac{a_n-ib_n}{2}=\frac{1}{T}\int_{-T/2}^{T/2}x(t)e^{-2\pi if_0nt}dt
$$
内積を
$$
<\phi,\psi>=\frac{1}{T}\int_{-T/2}^{T/2}\phi(t)\overline{\psi(t)}dt
$$
とする。$\phi(t)=e^{2\pi if_0nt}$とおくと、$\{\phi_n \}$は正規直交系となる。  
$$
||\phi_n||^2=\frac{1}{T}\int_{-T/2}^{T/2}e^{2\pi if_0nt}\overline{e^{2\pi if_0nt}}dt=\frac{1}{T}\int_{-T/2}^{T/2}dt=1
$$
となるので、$L^2$ノルムは1となり、$m\neq n$に対しては
$$
\begin{align}
<\phi_m,\phi_n>&=\frac{1}{T}\int_{-T/2}^{T/2}e^{2\pi if_0mt}e^{-2\pi if_0nt}=\frac{1}{T}\int_{-T/2}^{T/2}e^{2\pi if_0(m-n)t}\\
&=\biggl[\frac{e^{2\pi if_0(m-n)t}}{2\pi if_0(m-n)} \biggr]_{-T/2}^{T/2}=\frac{e^{(m-n)\pi i}-e^{-(m-n)\pi i}}{2\pi if_0(m-n)}\\
&=\frac{\sin{(m-n)\pi}}{\pi f_0(m-n)}=0
\end{align}
$$
となる。

$c_n$は複素フーリエ係数と呼ばれ
$$
c_n=<x,\phi_n>=\frac{1}{T}\int_{-T/2}^{T/2}x(t)e^{-2\pi if_0nt}dt
$$
と表せる。

$$
a_0=2c_0$、$a_n=2Re(c_n)\ (n\neq 0)$、$b_n=-2Im(c_n)$
$$
とすれば、実フーリエ係数が得られる。

パーセバルの等式は、
$$
|x|^2=\sum_{n=-\infty}^{\infty}|<x,\phi_n>|^2=\sum_{n=-\infty}^{\infty}|c_n|^2
$$
となり、シンプルなものになる。

(例1)
$$
x_1(t)=|t|(-\pi\leq t \leq \pi),\ x(t+2\pi)=x(t)
$$
という周期$T=2\pi$の関数を考える。
複素フーリエ係数は、$n\geq 1$のとき、
$$
\begin{align}
c_n&=\frac{1}{2\pi}\int_{-\pi}^{\pi}|t|e^{-int}dt=\frac{1}{2\pi}\int_{-\pi}^{\pi}|t|(\cos nt-i\sin nt)dt=\frac{1}{\pi}\int_{0}^{\pi}t\cos ntdt\\
&=\frac{1}{\pi}\biggl[t\frac{\sin nt}{n} \biggr]_0^\pi-\frac{1}{\pi}\int_{0}^{\pi}\frac{\sin nt}{n}dt=-\frac{1}{\pi}\biggl[t\frac{\cos nt}{n^2} \biggr]_0^\pi=-\frac{1-(-1)^n}{\pi n^2}
\end{align}
$$
$n=0$のとき、
$$
c_0=\frac{1}{2\pi}\int_{-\pi}^{\pi}|t|dt=\frac{1}{\pi}\int_{0}^{\pi}tdt=\frac{1}{\pi}\biggl[\frac{t^2}{2} \biggr]_0^\pi=\frac{\pi}{2}
$$
$x_1(t)$は次のように展開できる。
$$
x_1(t)\sim\frac{\pi}{2} - \sum_{n= -\infty,n\neq 0}^{\infty}\frac{1-(-1)^n}{\pi n^2}e^{in\pi t}
$$
対応する実フーリエ係数は、
$$
a_0=\frac{\pi}{2}\\
a_n=2Re\biggl\{\frac{1-(-1)^n}{\pi n^2} \biggr\}=\frac{2\{1-(-1)^n\}}{\pi n^2}\ (n\geq 1)\\
b_n=-2Im\biggl\{\frac{1-(-1)^n}{\pi n^2} \biggr\}=0\ (n\geq 1)
$$
$b_n=0$は$x(t)$が偶関数であることに対応する。

(例2)
$$
x_2(t)=e^{at}\ (-\pi\leq t < \pi),\ x_2(t+2\pi)=x_2(t)
$$
を考える。
$$
\begin{align}
c_n&=\frac{1}{2\pi}\int_{-\pi}^{\pi}e^{at}e^{-int}dt=\frac{1}{2\pi}\int_{-\pi}^{\pi}e^{(a-in)t}dt=\frac{1}{2\pi}\biggl[\frac{1}{a-in}e^{(a-in)t} \biggr]_{-\pi}^\pi\\
&=\frac{1}{2\pi}\frac{1}{a-in}(e^{(a-in)\pi}-e^{-(a-in)\pi})=\frac{1}{2\pi}\frac{(-1)^n}{a-in}(e^{a\pi}-e^{-a\pi})\\
&=\frac{\sinh{(ah)}}{\pi}\frac{(-1)^n}{a-in}=\frac{\sinh{(ah)}}{\pi}\frac{(-1)^n}{a^2+\pi^2}(a+in)
\end{align}
$$
対応する実フーリエ係数は、
$$
a_0=2c_0=\frac{2\sinh{(a\pi)}}{a\pi}\\
a_n=2Re(c_n)=\frac{2\sinh{(ah)}}{\pi}\frac{a(-1)^n}{a^2+\pi^2}\\
b_n=-2Im(c_n)=-\frac{2\sinh{(ah)}}{\pi}\frac{n(-1)^n}{a^2+\pi^2}
$$
