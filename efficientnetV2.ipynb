### EfficientNetV2

#### Fused-MBConv

depthwise convolutionは学習の効率を下げる要因となっていました。
特に入力に近い層ほど遅くしていたようです。
そこで、fused-MBConvが提案されました。MBConvの1×1畳み込み層とdepthwise convolutionを1つの3×3畳み込み層で置き換えたものです。
すべての層をfused-MBConvに置き換えるのではなく、最初の3ステージのみ入れ替えることで小さなコスト(パラメータ、FLOPS)で学習速度を改善することができました。



EfficientNetV1では一様にパラメータをスケールアップしていましたが、EfficientNetV2では後のステージほどより多くの層が追加されるようにします。
また、画像のサイズには最大値など制限を与えます。

#### 構造
EfficientNetV2-Sの構造を示します。

EfficientNetV1との違いは、

- MBConvとFused-MBConv(初めの3層)を使用
- expantion rateが小さくなっているものがある(表のMBconv〇の〇の部分)
- kernel sizeが3×3とサイズが小さくなっている(5×5がなくなっている)
- 最後の1ステージをなくしている



EfficientNEtV2-M/LへのスケールアップはEfficientNetV1と同じ考え方ですが、
- 画像のスケールの最大値を480とする
- 後の層ほど追加する層を増やす

という条件を追加します。
