### 撹乱項に関する標準的な過程
VARモデルの撹乱項$\boldsymbol{u}_t$に関する標準的な仮定は、次の形で表現されることが多い。
$$
\begin{align}
E(\boldsymbol{u}_t)&=0,\ t=1,2,\cdots,T\\
Var(\boldsymbol{u}_t)&=E(\boldsymbol{u}_t\boldsymbol{u}_t')\\
&=\boldsymbol{\Sigma}_u\\
&= \left(\begin{array}{cccc}
\sigma_{11}&\sigma_{12},\cdots,\sigma_{1K}\\
\sigma_{21}&\sigma_{22},\cdots,\sigma_{2K}\\
\vdots&\vdots,\cdots,\vdots\\
\sigma_{K1}&\sigma_{K2},\cdots,\sigma_{KK}
\end{array}\right)\\
t=1,2,\cdots,T\\
Cov(\boldsymbol{u}_t,\boldsymbol{u}_s)&=E(\boldsymbol{u}_t\boldsymbol{u}_s')=0,\ t\neq s
\end{align}
$$

2つ目の式は均一分散の多変量版を表し、時間の変化に対して共分散行列$\boldsymbol{\Sigma}_u$が均一であることを表している。
3つ目の式は撹乱項$\boldsymbol{u}_t$が系列無相関であることを表している。
撹乱項にかんする標準的な仮定として次の3つを調べる。
1. 撹乱項は系列無相関(自己無相関)である
2. 撹乱項は「均一分散」の形で分布している
3. 撹乱項は正規分布している

撹乱項の推定値である残差を用い、これらの仮定が満たされているか否かを調べる。

次の2ステップを踏まえてラグ次数$p$を決定することが考えられる。
1. 情報量基準に基づきラグ次数の候補を少数個に絞る
2. 少数個に絞ったラグ次数の候補の中から、撹乱項に関する基本的な仮定と整合的なラグ次数を選ぶ。撹乱項に関する基本的な仮定としては前述の3つが標準的である。

### 系列無相関の仮説検定
このタイプの検定では、ダービン=ワトソン検定がよく知られているが、VARにおけるラグ付き内生変数の存在やラグ次数$\rho$の一般化を含め、実用性の視点から、次の4つに絞って考えればよい。
1. Box-Piece検定(Portmanteau検定、カバン検定)
2. Ljung-Box検定(Adjusted Portmanteau検定)
3. Breusch-Godfrey LM検定
4. Edgerton-Shukur F検定

$K$次元ベクトルを対象とするPortmanteau検定(Box-Piece検定)の公式は次のようになる。

$$
PT=T\biggl\{\sum_{j=1}^htr(\boldsymbol{\hat{C}}_j'\boldsymbol{\hat{C}}_0^{-1}\boldsymbol{\hat{C}}_j\boldsymbol{\hat{C}}_0^{-1}) \biggr\}\sim_a \chi^2(K^2(h-p))
$$
ただし、
$$
\boldsymbol{\hat{C}}_j=\frac{1}{T}\sum_{t=j+1}^T\boldsymbol{\hat{u}}_t\boldsymbol{\hat{u}}_{t-j}'\\
\boldsymbol{\hat{u}}_tは最小二乗推定の残差
$$
である。
ラグ次数$h$は、モデル用ラグ次数$p$よりも十分に大きく選ぶことが求められる。
この検定統計量は次のように解釈できる。$\boldsymbol{\hat{C}}_j$がラグ次数$j$の自己相関係数であり、$j=1,2,\cdots,h$といったラグ次数の自己相関係数を調べ、総合的に残差の系列相関を計測している。
帰無仮説が真であるとき、理屈的には現実と帰無仮説のズレ$\boldsymbol{\hat{C}}_j$がゼロとなり、検定統計量$PT$もゼロになる。
したがってゼロを基点とする右片側検定になる。
小標本向けのPrtmanteau検定(Ljung-Box検定)では次のようになる。
$$
PT_{adj}=T\biggl\{\sum_{j=1}^h\frac{T}{T-j}tr(\boldsymbol{\hat{C}}_j'\boldsymbol{\hat{C}}_0^{-1}\boldsymbol{\hat{C}}_j\boldsymbol{\hat{C}}_0^{-1}) \biggr\}\sim_a \chi^2(K^2(h-p))
$$
一方、LM検定となるBreusch-Godfrey LM検定は次の補助回帰式を用いる。
$$
\boldsymbol{\hat{u}}_t=\sum_{j=1}^p\boldsymbol{B}_j\boldsymbol{y}_{t-i}+\boldsymbol{B}_z\boldsymbol{z}_t+\sum_{j=1}^h\boldsymbol{D}_j\boldsymbol{\hat{u}}_{t-j}+\boldsymbol{v}_t
$$
ただし、$\boldsymbol{v}_t$はホワイトノイズであり、$\boldsymbol{D}_j(j=1,2,\cdots,h)$は$K×K$次元の係数行列であり、$h$は検定に用いるラグ次数である。
帰無仮説は次のようになる。
$$
H_0:\boldsymbol{D}_1=\boldsymbol{D}_2=\cdots=\boldsymbol{D}_h=\boldsymbol{0}
$$
制約の個数は$hK^2$になる。そしてLM検定の公式は次のように表現できる。
$$
LM_{BG}=T\{KR^2_{per}\}\sim_a\chi^2(hK^2)
$$
ただし、
$$
R_{per}^2=1-\frac{1}{K}tr\bigl[\boldsymbol{\hat{\Sigma}}_{ax}(H_*)\boldsymbol{\hat{\Sigma}}^{-1}_{ax}(H_0) \bigr]
$$
である。
$R_{per}^2$は$K$次元からなる補助回帰式における方程式1本あたりの平均的な決定係数を表す。
$\boldsymbol{\hat{\Sigma}}_{ax}(H_0)$は補助回帰式において帰無仮説$H_0$を真とする制約を加えたときの制約付き推定共分散行列、$\boldsymbol{\hat{\Sigma}}_{ax}(H_*)$は制約なし推定共分散行列を表す。
帰無仮説が真であるとき、理屈的には$R_{per}^2$がゼロとなり、$LM_{BG}$もゼロになる。したがってゼロを基点とする右片側検定になる。
Breusch-Godfrey LM検定に対して小標本向けとなるEdgerton-Shukur F検定は次のようになる。
$$
LM_{ES}=\frac{1-(1-R_{all}^2)^{\frac{1}{r}}}{(1-R_{all}^2)^{\frac{1}{r}}}\frac{Nr-q}{Km}\sim F[hK^2,int(Nr-q)
$$
ただし、
$$
\begin{align}
R_{all}^2&=1-\frac{det(\hat{\boldsymbol{\Sigma}}_{ax}(H_*))}{det(\hat{\boldsymbol{\Sigma}}_{ax}(H_0))}\\
m&=Kh\\
r&=\biggl(\frac{K^2m^2-4}{K^2+m^2-5} \biggr)^{\frac{1}{2}}\\
q&=\frac{1}{2}Km-1\\
N&=T-K-m-\frac{1}{2}(K-m+1)
\end{align}
$$
である。
検定統計量$LM_{ES}$は近似的にF分布に従う。
帰無仮説(系列無相関の仮説)が真であるとき、理屈的には$R_{all}^2$がゼロとなり、$LM_{ES}$もゼロになる。
したがってゼロを基点とする右片側検定になる。

LM検定統計量の見方について説明する。
1本の方程式からなる回帰モデルにおいて、係数の線形規約に関する帰無仮説を$H_0$を考える。対立仮説$H_1$は制約なしを意味する仮説になる。
このLM検定統計量は次の公式になることが知られている。
$$
\begin{align}
LM&=T\biggl\{\frac{SSR(H_0)-SSR(H_*)}{SSR(H_0)} \biggr\}\\
&=T\biggl\{1-\frac{SSR(H_*)}{SSR(H_0)} \biggr\}\sim \chi^2(J)
\end{align}
$$
ただし、$SSR(H_0)$は制約付きモデルの残差変動、$SSR(H_*)$は制約なしモデルの残差変動、$J$は線形独立となっている制約の個数、$T$は標本サイズである。
制約付きモデルは帰無仮説$H_0$のもとでもモデルであることから、その残差変動($SSR$)は$SSR(H_*)$と表記している。
ここではモデル全体の有意性を計るため、すべての説明変数の係数がゼロであるとする帰無仮説を考える。
定数項付きの定式化を想定し、定数項は制約のない自由パラメータとする。この場合は、
$$
SSR(H_0)=SST(H_*)
$$
となる。
ただし、$SST(H_*)$は回帰の総変動($SST$)を表す。
したがって、次の表現を得ることができる。
$$
\begin{align}
LM&=T\biggl\{1-\frac{SSR(H_*)}{SSR(H_0)} \biggr\}\\
&=T\biggl\{1-\frac{SSR(H_*)}{SST(H_*)} \biggr\}\\
&=T×R^2\sim \chi^2(J)
\end{align}
$$
ただし、$R^2$は回帰の決定係数であり、
$$
R^2=1-\frac{SSR(H_*)}{SST(H_*)}
$$
と定義される。

### 均一分散の仮説検定
回帰分析においても、VAR分析においても、撹乱項が均一分散になっていることは標準的な仮定である。
VARにおける均一分散は、時間の変化に対して共分散行列が均一であるといった内容になる。
VAR分析に限定するならば、次の3つに絞って考えればよい。

1. Breusch-Pagan検定(Breusch-Pagan-Godfrey検定)
2. White検定
3. ARCH-LM検定

これら3つの検定はラグランジュ乗数(LM)検定となる。
まず、単純化(一変量版)によってBreusch-Pagan検定の考え方を示す。不均一分散の構造として次のような構造を考える。
$$
\sigma_t^2=\delta_0+\boldsymbol{\delta}_1'\boldsymbol{x}_t+v_t,\hspace{3mm}v_t\sim WN(0,\sigma^2)
$$
ただし、$\sigma_t^2$は$t$期における撹乱項の分散、$\boldsymbol{x}_t$は不均一分散の構造を特定する変数ベクトル、$v_t$はホワイトノイズである。
帰無仮説と対立仮説は次の内容になる。

$$
H_0:\boldsymbol{\delta}_1=\boldsymbol{0}(均一分散)\\
H_1:\boldsymbol{\delta}_1\neq\boldsymbol{0}(不均一分散)
$$

実際には未知な分散$\sigma_t^2$に代えて残差2乗$\hat{u}_t^2$を用い、次の補助回帰式を利用する。
$$
\hat{u}_t^2=\delta_0+\boldsymbol{\delta}_1'\boldsymbol{x}_t+v_t,\hspace{3mm}v_t\sim WN(0,\sigma^2)
$$

Breusch-Pagan検定は、不均一分散の構造を特定する変数$\boldsymbol{x}_t$を選び出す知識が必要である。

一方、White検定は回帰式にも含まれているすべての説明変数と共に、それらの2乗項やクロス積を$\boldsymbol{x}_t$として用いる点が特徴的である。
White検定は不均一分散の構造を特定しないため、Breusch-Pagan検定よりも一般的な場合に使える。
しかし、不均一分散の構造を特定しないため、帰無仮説(均一分散の仮説)が棄却されても、不均一分散に対する次の対策が立てられない。

これに対し、ARCH-LM検定は残差2乗の過去値を$\boldsymbol{x}_t$として用いる点が特徴的である。
したがって、不均一分散の構造を特定する変数を選び出す知識を必要としない。
このような検定の実用性を考慮すると、VARモデルにおける均一分散の検定はARCH-LM検定に絞ってよい。
ただし、VARモデルであるので、ARCH-LM検定の多変量版(multivariate ARCH-LM)を用いる。

ARCH-LM検定の多変量版(multivariate ARCH-LM)では次の補助回帰式を用いる。
$$
vech(\hat{\boldsymbol{u}}_t,\hat{\boldsymbol{u}}_t')=\boldsymbol{d}_0+\sum_{j=1}^q\boldsymbol{D}_jvech(\hat{\boldsymbol{u}}_t,\hat{\boldsymbol{u}}_{t-j})+\boldsymbol{v}_t
$$
この補助回帰式に示す$\boldsymbol{v}_t$はホワイトノイズ・ベクトルであり、$q$は検定のためのラグ次数を表す。
$vech(・)$は正方行列の列を取り出してベクトルへ変換するオペレータであり、対角要素と、その下側の要素を取り出す。
$\hat{\boldsymbol{\Sigma}}=\hat{\boldsymbol{u}}_t\hat{\boldsymbol{u}}_t'$が$K×K$の対称行列であるから、$vech(\hat{\boldsymbol{u}}_t\hat{\boldsymbol{u}}_t')$は$\frac{1}{2}K(K+1)$次元のベクトルになる。
これを受け、$\boldsymbol{d}_0$は$\frac{1}{2}K(K+1)$次元の定数項ベクトル、$\boldsymbol{D}_j(j=1,2,\cdots,q)$は$\frac{1}{2}K(K+1)×\frac{1}{2}K(K+1)$の係数行列である。
そして帰無仮説(均一分散の仮説)は次のようになる。
$$
H_0:\boldsymbol{D}_1=\boldsymbol{D}_2=\cdots=\boldsymbol{D}_q=\boldsymbol{0}
$$
制約の個数は$\frac{qK^2(K+1)^2}{4}$となる。
そして検定の公式は次のように表現できる。
$$
LM_{ARCH}=T\biggl\{\frac{1}{2}K(K+1)R_{per}^2 \biggr\}\sim \chi-2\biggl\{\frac{qK^2(K+1)^2}{4} \biggr\}
$$
ただし、
$$
R_{per}^2=1-\frac{2}{K(K+1)}tr\bigl[\hat{\boldsymbol{\Sigma}}_{ax}(H_*)\hat{\boldsymbol{\Sigma}}_{ax}^{-1}(H_0) \bigr]
$$
である。
$R_{per}^2$は$\frac{1}{2}K(K+1)$次元の補助回帰式における方程式1本あたりの平均的な決定係数を表す。
帰無仮説が真であるとき、理屈的には$R_{per}^2$がゼロとなり、$LM_{ARCH}$もゼロになる。
したがって、ゼロを基点とする右片側検定になる。

### 正規分布の仮説検定
撹乱項の分布としては正規分布を仮定するのが標準的である。
その妥当性を調べる検定として、ジャック=ベラ検定がよく知られている。
正規分布は歪度と尖度とがゼロである性質があり、それを利用するのがジャック=ベラ検定である。
$K$次元ベクトルに基づくジャック=ベラ検定の公式は次のようになる。
$$
JB_K=T\biggl(\frac{1}{6}S_3^2+\frac{1}{24}S_4^2 \biggr)\sim\chi^2(2K)
$$
ただし、
$$
S_3^2=\boldsymbol{b}_3'\boldsymbol{b}_3\\
S_4^2=(\boldsymbol{b}_4-\boldsymbol{3}_K)'(\boldsymbol{b}_4-\boldsymbol{3}_K)
$$
である。
$\boldsymbol{b}_3$は標準化された残差に基づく3次モーメント($K$次元ベクトル)、$\boldsymbol{b}_4$は標準化された残差に基づく4次モーメント($K$次元ベクトル)である。
正規分布由来の標本歪度は期待値が0、標本尖度は期待値が3であることから、$S_4^2$のほうはゼロになるような中心化調整を行っている。
つまり$\boldsymbol{3}_K=(3,3,\cdots,3)$からなる$K$次元ベクトルを用い、$\boldsymbol{b}_4-\boldsymbol{3}_K$という処理を行っている。
そして標準化された残差は、
$$
\hat{\varepsilon}^{SD}_t=\boldsymbol{P}^{-1}(\hat{\boldsymbol{u}}_t-\bar{\hat{\boldsymbol{u}}}_t)
$$
で与えられている。
$\boldsymbol{P}$は推定共分散行列$\hat{\boldsymbol{\Sigma}}_u$のコレスキー分解行列を表す。
コレスキー分解行列$\boldsymbol{P}$は$\hat{\boldsymbol{\Sigma}}_u=\boldsymbol{P}\boldsymbol{P}'$といった性質を持つ下三角行列である。
帰無仮説(正規分布の仮説)が真であるとき、理屈的にはデータが示す歪度$S_3^2$も、尖度$S_4^2$もゼロとなり、検定統計量$JB_K$もゼロを示す。
したがって、ゼロを基点とする右片側検定になる。
